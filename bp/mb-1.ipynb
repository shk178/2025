# 비트코인, 공개 블록체인 프로그래밍 (안드레아스)
## 1. 서론
`그냥 쭉 읽는다. (실습x, 검색x)`

### p.39
> - 비트코인의 구성요소는 다음과 같다.
>   - 분산화된 P2P 네트워크(비트코인 프로토콜)
>   - 공개거래장부(블록체인)
>   - 분산화된 수학적 결정론적 통화 발행(분산 채굴)
>   - 분산화된 거래 검증 시스템(거래 스크립트)
> - 나는 개발자로서 비트코인을 인터넷머니와 유사하며 분산 계산을 통해 화폐 가치를 보급하고 디지털 자산의 소유권을 안전하게 보호하는 네트워크라고 생각한다.

### p.42-p.43
> - **북미 지역의 저가 소매**
>   - 이 책에서는 앨리스가 비트코인에 대해 공부하고 구매한 후 팔로알토에 있는 밥스카페에서 커피 한 잔을 비트코인으로 구매하기까지의 이야기를 소개할 것이다.
>   - 이 일화를 통해 우리는 소매 고객의 관점에서 비트코인 소프트웨어와 비트코인의 교환 및 기본 거래에 대해서 살펴볼 것이다.
> - **북미 지역의 고가 소매**
>   - 캐롤은 비트코인을 받고 고가의 작품을 판매하고 있다.
>   - 이 일화를 통해 고가인 물건을 거래하는 소매업자들에게 발생할 수 있는 '51%' 어택의 위험성을 다룰 것이다.
> - **해외계약 서비스**
>   - 밥은 인도의 방갈로르에 거주하고 있는 웹 개발자인 조셉과 거래했다.
>   - 이 일화를 통해 아웃소싱, 계약서비스, 해외송금 등에서 비트코인이 어떻게 사용되는지에 대해 알아볼 것이다.
> - **인터넷 쇼핑몰**
>   - 가브리엘은 리우데자네이루에 사는 진취적인 10대 소년으로 비트코인 상표를 붙인 티셔츠, 커피 머그잔, 스티커 등을 판매하는 소규모 인터넷 쇼핑몰을 운영하고 있다.
> - **자선 기부금**
>   - 유지니아는 필리핀에 있는 아동 자선단체 본부장이다.
>   - 이 일화를 통해 국경과 통화를 초월하여 전 세계적인 모금 활동을 위해 비트코인이 사용되는 사례와 자선단체 내에서 투명성 제고를 위해 공개장부를 사용하는 사례에 대해 알아볼 것이다.
> - **수입/수출**
>   - 모하메드는 현재 수입품에 대한 대금 지불을 신속하게 진행하기 위해 미국과 중국에서 아랍에미레이트로 들어오는 전자장비 수입품을 비트코인으로 구매하려 한다.
>   - 이 일화에서는 물품에 대한 기업 간 대규모 해외지불에 비트코인이 어떻게 사용될 수 있는지를 보여준다.
> - **비트코인 채굴**
>   - 징은 공학기술을 이용하여 수입을 늘릴 목적으로 비트코인을 채굴하는 '채굴' 리그를 만들었다.
>   - 이 일화에서는 비트코인의 '산업' 기반에 대해 점검해 볼 것이다.
>   - 즉, 비트코인 네트워크를 안전하게 보호하면서 새로운 통화를 발행하는 데 사용되는 전문 장비에 대해 알아본다.
