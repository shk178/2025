1. 데이터 이해
`데이터, 정보`
- 데이터 정의
    - 데이터: 객관적 사실, 가공 전 (주문수량)
    - 정보: 데이터로부터 가공된 자료 (베스트셀러)
- 데이터 유형
    - 정성적/정량적 데이터
        - 정성적: 자료 수치화/수치, 기호(온도, 풍속)
        - 정량적: 자료 특징 풀어 설명/언어, 문자 (기상특보, 주관식 설문 응답)
    - 정형/반정형/비정형 데이터
        - 정형: 정보 형태가 정해짐 (관계형db, 엑셀, csv)
        - 반정형: 데이터 설명하는 메타데이터 포함 (html, xml, json, rdf)
        - 비정형: 형태 정해지지 않은 정보 (sns, 유튜브, 음원)
- 암묵지-형식지 상호작용
    - 암묵지: 개인에게 습득되고 겉으로 드러나지 않는 지식
    - 형식지: 문서, 메뉴얼 등의 형상화된 지식
    - 상호작용
        - 공통화: 암묵지 지식을 타인에게 알려줌
        - 표출화: 암묵지 지식을 문서메뉴얼로 전환
        - 연결화: 교재메뉴얼에 새로운 지식 추가
        - 내면화: 교재메뉴얼에서 타인의 암묵지 터득
- dikw 피라미드
    - data데이터: 있는 그대로 사실 (a대리점 100만원, b대리점 200만원)
    - information정보: 데이터 통한 패턴 인식 (a대리점이 싸다)
    - knowledge지식: 패턴 통한 예측 (a대리점에서 사는 게 이득일 거다)
    - wisdom지혜: 창의적인 산물 (a대리점의 다른 제품들도 쌀 거다)
- 데이터 단위
    - kb<mb<gb<tb: 2의 10승 20승 30승 40승
    - pb<eb<zb<yb: 2의 50승 60승 70승 80승 (페타 엑사 제타 요타바이트)

`데베 정의와 특징`
- 데베 개념
    - db: 구조조직화된 data 집합
        - 스키마: db 구조와 제약조건관한 명세 (외부 개념 내부스키마)
        - 인스턴스: 데이터 개체 구성하는 속성에 대한 데이터 타입과 값
        - 메타데이터: 데이터 설명용 데이터, 데이터 구조 설명하고 검색 시 활용
        - 인덱스: 정렬 탐색 위한 데이터명
    - dbms: db관리, 접근 제공하는 sw
        - 관계형 dbms: 표로 정리 (mysql, mariadb, oracle)
        - nosql dbms: 비정형 데이터 저장 처리 (hbase, mongodb, couchbd, redis, cassandre)
    - sql: 데베 접근하는 하부언어
        - ddl정의언어: create, alter, drop
        - dml조작언어: select, insert, delete, update
        - dcl제어언어: commit, rollback, grant, revoke
- 데이터베이스 특징
    - 공용 데이터다.: 여러 사용자가 다른 목적으로 데이터 공동 이용한다.
    - 통합된 데이터다.: 동일한 데이터 중복되지 않는다.
    - 저장된 데이터다.: 저장매체에 저장한다.
    - 변화되는 데이터다.: 새로운 데이터 추가/수정/삭제 가능하고 현재의 정확한 데이터도 유지한다.
- 데베 설계 절차
    - 먼저 요구조건 분석한다.
    - 개념적 설계한다 (개념 스키마)
    - 논리적 설계한다 (개념 erd를 활용한 논리적 모델링)
    - 물리적 설계한다 (저장 구조 설계한다)

`데이터 활용`
- 기업 활용 데베
    - oltp: 데이터를 수시로 갱신 (거래단위)
    - olap: 다차원 데이터를 대화식으로 분석
    - crm: 고객과 관련 자료 분석/마케팅 활용
    - scm: 공급망 연결 최적화
    - erp: 기업 경영 자원을 효율화
    - rte: 최신 정보로 빠른 의사결정 지원
    - bi: 기업 보유 데이터 정리분석하는 리포트 도구
    - ba: 통계 기반 비즈니스 통찰력
    - blockchain: 네트워크 참여 사용자가 정보 분산저장
    - kms: 기업의 모든 지식 포함
- dw데이터웨어하우스
    - 특징
        - 주체지향성: 분석목적 설정중요
        - 데이터통합: 일관화된 형식저장
        - 시계열성: 히스토리있는 데이터
        - 비휘발성: 읽기전용 (수시변화x)
    - 구성요소
        - etl: extraction, transform, load
        - ods (operational data store): 다양한 dbms에서 추출한 데이터 임시 저장
- datalake데이터레이크
    - 비정형 데이터를 저장하며 하둡과 연계해 처리
    - 하둡: 여러 컴퓨터 하나로 묶어 대용량 데이터 처리하는 오픈소스 빅데이터 솔루션
        - hdfs: 분산형 파일 저장 시스템
        - mapreduce: 분산된 데이터 병렬 처리

2. 데이터의 가치와 미래
`빅데이터 이해`
- 빅데이터 출현 배경
    - 인터넷 확산 스마트폰 보급
    - 클라우딩 컴퓨터로 경제성 확보
    - 저장매체 가격하락
    - 하둡으로 분산 컴퓨팅
    - 비정형 데이터 확산
- 3v (가트너 정의)
    - volume규모: 데이터량 증가 (구글 번역)
    - variety다양성: 데이터유형 다양화
    - velocity속도: 데이터생성처리 속도증가
    - 5v/7v
        - value: 숨겨진가치발견
        - veracity신뢰성: 고품질데이터
        - validity정확성: 데이터유효성보장
        - volatility휘발성: 데이터의미있는기간
- 빅데이터 비유
    - 산업혁명의 석탄과 철 같다.
    - 원유 같다: 정보제공해서 생산성늘린다.
    - 렌즈같다: 구글 ngram viewer, 현미경이 생물학발전시켰듯, 산업전반영향주었다
    - 플랫폼: 공동활용목적으로 구축된 구조물이다. 써드파티(원천기술 활용한 파생상품 만드는 회사) 비즈니스에 활용된다. (페이스북)
- 빅데이터가 만드는 변화
    - 표본조사 ->전수조사
    - 사전처리 ->사후처리
    - 질 ->양
    - 인과관계 ->상관관계

`비즈니스 모델`
- 빅데이터 활용위한 3대 요소
    - 인력 자원(데이터) 기술
- 빅데이터 주요분석기법
    - 회귀분석: 독립변수-종속변수 관계/x는 y에 어떤 영향 미치는가 (수도권 가까울수록 부동산 비싼가)
    - 분류분석: a와 b가 속하는 범주 어디인가 (고양이 강아지 이미지 판별)
    - 연관규칙: 여러 요소들간 규칙 상관관계 존재하는가 (마트에서 치킨과맥주 같이삼)
    - 유전자알고리즘: 최적화 필요한 문제의 해결책 (택배차량 어떻게 배치, 최대시청률 얻으려면 어떤 프로그램을 어떤 시간대에 방송)
    - 기계학습: 훈련데이터로부터 컴퓨터가 학습하고 미래를 예측 (넷플릭스 추천)
    - 감정분석: 텍스트 데이터에서 감정(긍부정) 분석
    - sns분석: 사람간관계 (sns상 사용자들 관계 속 영향력 높은/낮은 사람)
    - 텍스트마이닝: 텍스트로부터 자연어처리nlp 통한 숨겨진 의미 발견 (문서요약 키워드추출)

`위기요인과 통제방안`
- 위기요인과 통제방안
    - 사생활침해: sns데이터가 사생활 침해/제공자에서 침해한 사용자책임으로 전환
    - 책임원칙훼손: 범죄예측프로그램으로 예측해 체포하는 문제/결과 대해서만 책임
    - 데이터오용: 분석결과가 항상옳지는 않음/알고리즘 해석가능한 알고리즈미스트(부당하게 피해 발생한 사람들 구제하는 전문인력) 필요
- 데이터 3법
    - 가명정보 개념 도입 (통계작성 연구 공익적기록보존 목적 하에 동의없이 활용 가능)
    - 개인정보보호법
    - 정보통신망 이용 촉진 및 정보보호 등에 관한 법률(정보통신망법)
    - 신용정보 이용 및 보호에 관한 법률 (신용정보법)
- 개인정보, 가명정보, 익명정보
    - 개인정보: 개인을 알아볼 수 있는 정보/동의 받아 활용 가능 (홍길동, 33세)
    - 가명정보: 가명처리로 특정불가 (홍00, 30대 초반)
    - 익명정보: 더이상 개인 알아볼 수 없는 정보/제한 없이 활용 가능 (000, 30대)
- 개인정보 비식별화
    - 가명처리 (홍길동 35세 -> 임꺽정 30세)
    - 총계처리 (홍길동 170, 임꺽정 180 -> 평균키 175)
    - 데이터 처리 (주민번호 901111-1xxxxxx -> 90년대 생, 남)
    - 데이터 범주화 (홍길동 35세 -> 홍길동 30~40세)
    - 데이터 마스킹 (홍길동 35세 -> 홍00 35세)
- 프라이버시 보호 모델
    - k-익명성: 같은 값이 존재하도록 해서 다른 정보로 결합 불가토록
    - l-다양성: 민감한 정보의 다양성 높여 추론 가능성 낮춤
    - t-근접성: 민감 정보의 분포 낮추어 추론 가능성 낮춤

`미래의 빅데이터`
- 데이터 산업의 발전
    - 처리->통합->분석->연결->권리
        - 처리: 프밍언어 통한 데이터처리
        - 통합: dbms등장
        - 분석: 빅데분석기술발전
        - 연결: api활용한 모듈연결
        - 권리: mydata활용한 데이터 주권 행사
            - 마이데이터: 자기 신용정보를 제3자에게 제공해 서비스 제공받음

3. 가치 창조 위한 데이터 사이언스와 전략 인사이트
`빅데이터 분석과 전략 인사이트`
- 전략인사이트
    - 집중과선택 (많은데이터나 다양한대상을 분석하기보다 현재분석에 집중)
    - 업계상황만 보지말고 더 넓은 시야로 봐야됨
    - 경영진의 전략적 인사이트에 기여
    - "조직이분석 배우는중 혹은 특정문제범위 해결시": 집중과 선택
    - "사업상황 확인시": 넓은시야
- 데이터사이언스
    - 데이터 관련된 모든 전문지식 종합학문
    - 정형/비정형 데이터 모두 분석 (총체적접근)
- 데사 핵심구성요소
    - analytics: 이론지식
    - it: 프밍지식
    - 비즈니스분석: 비즈니스능력

`전략 인사이트 도출위한 필요역량`
- 데이터 사이언티스트 필요역량
    - 하드스킬: 이론지식(수학/통계/가설검정), 가트너제시역량에 미포함
    - 소프트스킬: 스토리텔링, 리더십, 창의력, 분석
    - 하드스킬 이과적/소프트스킬 문과적

`데이터 그리고 데사 미래`
- 빅데이터 가치 패러다임 변화한다.
    - digitalization->connection->agency
    - 디지털화: 아날로그세상디지털화
    - 연결: 정보연결
    - agency: 연결 관리
