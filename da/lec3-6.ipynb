- 회귀 모형의 검정
    - 독립변수와 종속변수 설정
    - 회귀계수 값의 추정
    - 모형이 통계적으로 유의미한가
        - 모형에 대한 F 통계량, p-value
        - 귀무가설: "모든 회귀계수는 0이다"
    - 회귀계수들이 유의미한가
        - 회귀계수들의 t 통계량, p-value
        - 각각의 회귀계수에 대한 귀무가설: "회귀계수는 0이다"
    - 귀무가설이 모두 기각하면 해당 모델을 활용
    - 모형이 설명력을 갖는가
        - 결정계수 R-square 값
- 회귀 모형 검정은 주어진 데이터를 바탕으로 변수들 간의 관계를 찾고, 그 관계가 통계적으로 신뢰할 만한지를 평가하는 과정입니다.
    - 독립변수와 종속변수 설정
        - 독립변수(X)는 원인이 되는 변수, 종속변수(Y)는 결과가 되는 변수입니다.
    - 회귀계수 값의 추정
        - 회귀분석을 수행하면 X와 Y의 관계를 나타내는 계수(기울기와 절편)가 계산됩니다.
    - 모형이 통계적으로 유의미한가?
        - F 통계량과 p-value를 통해 회귀 모형 자체가 의미 있는지를 평가합니다.
        - 귀무가설: "모든 회귀계수는 0이다" → 즉, X와 Y가 관련 없다는 뜻입니다.
        - p-value가 작으면(보통 0.05 이하) 귀무가설을 기각하고, 회귀모형이 의미 있다고 판단합니다.
    - 각 회귀계수가 유의미한가?
        - t 통계량과 p-value를 통해 각각의 독립변수가 종속변수에 영향을 미치는지를 확인합니다.
        - 귀무가설: "해당 회귀계수는 0이다" → 즉, 특정 X가 Y에 영향을 안 준다는 뜻입니다.
        - p-value가 작으면 그 독립변수가 유의미하다고 봅니다.
    - 모형이 설명력을 갖는가?
        - 결정계수(R-square)를 통해 모형이 데이터를 얼마나 잘 설명하는지를 평가합니다.
        - R-square 값이 클수록(0~1 범위) 모형이 데이터를 잘 설명하는 것입니다.
    - p-value를 보고 모형과 변수들이 유의미하면 활용할 수 있고, R-square 값이 높을수록 신뢰할 만한 모형이라 볼 수 있습니다!

- 일반적인 회귀 모형의 검정결과 해석
    - `Call:`
    - `lm(formula = height ~ age + no_sibillings, data = ageandheight)`
        - 회귀 분석을 실행한 코드입니다.
        - `height`를 종속변수로 두고, `age`와 `no_sibillings`를 독립변수로 사용했습니다.
    - `Residuals:`
    - `Min: -0.28029`
    - `1Q: -0.22490`
    - `Median: -0.02219`
    - `3Q: 0.14418`
    - `Max: 0.48350`
        - 잔차(Residuals)는 실제 값과 예측 값의 차이입니다.
        - 최소값과 최대값이 -0.28029에서 0.48350 사이로 분포하며, 중앙값은 -0.02219입니다.
        - 이는 예측이 실제 값과 비교적 가깝다는 뜻입니다.
    - `Coefficients:`
        - Coefficients (회귀 계수)
            - Estimate이 회귀 계수 예상값
        - 표준 오차 (Std. Error): 추정된 회귀 계수들이 얼마나 정확한지를 나타냅니다.
            - 작은 값일수록 신뢰성이 높습니다.
    - `(Intercept) Estimate: 64.95872`
    - `(Intercept) Std. Error: 0.55752`
    - `(Intercept) t value: 116.515`
    - `(Intercept) Pr(>|t|): 1.28e-15***`
        - `(Intercept)`의 값이 64.95872로, 나이와 형제 수가 0일 때 키의 예상값입니다.
    - `age Estimate: 0.63516`
    - `age Std. Error: 0.02254`
    - `age t value: 28.180`
    - `age Pr(>|t|): 4.34e-10***`
        - `age` 계수는 0.63516으로, 나이가 한 살 많아질 때 평균적으로 키가 0.63516 증가한다는 의미입니다.
        - `age`의 t-값이 28.180로 매우 크고, p-값이 4.34e-10로 거의 0에 가까워서 매우 유의미한 변수라는 뜻입니다. 
        - `age`는 유의성 코드가 `***`로 아주 중요한 변수임을 확인할 수 있습니다.
    - `no_sibillings Estimate: -0.01137`
    - `no_sibillings Std. Error: 0.05893`
    - `no_sibillings t value: -0.193`
    - `no_sibillings Pr(>|t|): 0.851`
        - `no_sibillings` 계수는 -0.01137로, 형제 수가 많아질수록 키에 미치는 영향이 거의 없음을 뜻합니다.
        - `no_sibillings`의 t-값은 -0.193이고 p-값이 0.851이라 유의미하지 않다는 것을 보여줍니다.
            - 형제 수는 키에 큰 영향을 미치지 않는다는 뜻입니다.
    - `Signif. codes:`
        - `***`은 매우 강한 유의성을, `**`, `*`, `.`은 상대적으로 낮은 유의성을 뜻합니다.
    - `0 '***' 0.001 '**' 0.01 '*' 0.05 '.' 0.1 ' ' 1`
    - `Residual standard error: 0.2693 on 9 degrees of freedom`
        - 모델이 얼마나 정확한지를 나타냅니다. 값이 작을수록 예측이 더 정확합니다.
    - `Multiple R-squared: 0.9888, Adjusted R-squared: 0.9863`
        - `Multiple R-squared: 0.9888`은 모델이 데이터의 98.88%를 설명한다는 뜻입니다.
        - `Adjusted R-squared: 0.9863`은 변수를 추가해도 모델이 여전히 높은 설명력을 가지는지를 나타냅니다.
    - `F-statistic: 397.7 on 2 and 9 DF, p-value: 1.658e-09`
        - DF= degree of freedom
        - F-값이 397.7로 매우 크고, p-값이 1.658e-09로 거의 0에 가까워서 전체적으로 모델이 매우 유의미하다는 것을 보여줍니다.
        - 나이는 키에 큰 영향을 미치지만, 형제 수는 거의 영향을 주지 않습니다.
- 종속변수 height
- 독립변수 age, no_sibillings
- 회귀모형 F분포의 p-value(1.658e-09)가 0.05보다 작으므로 모형이 유의미
    - e-09=1/(10^9)
- age의 p-value(4.34e-10)가 0.05보다 작으므로 회귀계수 유의미
- no_sibillings의 p-value(0.851)가 0.05보다 크므로 제외하고 회귀분석 재수행을 권장
- 위 모형은 다중회귀 모형
- R-square: 0.98888, 수정된 R제곱: 0.9863 (모형은 전체 데이터의 98% 이상을 설명)
- 회귀 자유도 2, 잔차의 자유도 9 -> 총 2+9+1=12개 데이터 활용한 분석
- 모델 회귀 식: Yheight=0.63516Xage-0.01137Xno_sibillings+64.95872
- no_sibillings 변수가 유의미하지 않기에 제거하고 검정 재수행은 연구자 판단
