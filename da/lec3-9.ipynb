3. 정형 데이터 마이닝
`데이터 마이닝 개요`
- 데이터 마이닝
    - 방대한 데이터 속에서 새로운 규칙, 패턴을 찾고 예측을 수행하는 분야
- 데이터 마이닝 유형
    - 지도학습: 정답이 있는 데이터를 활용
        - 인공신경망, 의사결정트리, 회귀분석, 로지스틱회귀
    - 비지도학습: 정답이 없는 데이터들 사이의 규칙을 파악
        - 군집분석, SOM, 차원축소, 연관분석
- 과대적합과 과소적합
    - 학습(50)<->검증(30)->평가(20)
    - 과대적합
        - 모델이 지나치게 데이터 학습해 매우 복잡해진 모델
        - 높은 분산, 낮은 편향
    - 과소적합
        - 데이터를 충분히 설명못하는 단순한 모델
        - 낮은 분산, 높은 편향
    - overfitting: 선이 복잡하게 구불거림
    - justfitting: 선이 적절히 구부러짐
    - underfitting: 선이 거의 일직선이어서 제대로 분류x
- 데이터 분할
    - 과대적합과 과소적합을 방지하고
    - 데이터가 불균형한 문제를 해결하기 위해 사용한다.
    - 분할된 데이터셋 종류
        - 훈련용 (training set): 모델 학습에 활용
        - 검증용 (validation set): 모델 과대과소 적합 조정에 활용
        - 평가용 (test set): 모델 평가에 활용
    - 분할된 데이터 학습 및 검증법
        - 홀드아웃
            - 훈련용과 평가용 2개 셋으로 분할
        - k-fold 교차검증
            - 데이터를 k개의 집단으로 구분해서
            - k-1개 학습, 나머지 1개로 평가
        - LOOCV
            - 1개 데이터로만 평가, 나머지로 학습
            - 데이터 수 부족시 적용
        - 부트스트래핑
            - 복원추출 활용해 데이터셋 생성
            - 데이터 부족불균형 문제 해결
            - 앙상블

`분류분석`
- 로지스틱 회귀분석
    - 종속변수가 범주형 데이터를 대상으로 2개 집단을 분류하는 문제에 활용
        - 데이터를 두 개 이상의 범주(예: 성공 vs 실패, 스팸 vs 정상 메일)로 나누는 데 사용되는 통계적 기법이에요.
        - 일반적인 선형 회귀(Linear Regression)는 숫자 값을 예측하지만, 로지스틱 회귀는 범주형 데이터를 분류하는 데 사용됩니다.
        - 특히 이진 분류(Binary Classification) 문제에서 많이 사용되며, 예를 들어 이메일이 스팸인지 아닌지를 분류하는 데 쓰일 수 있어요.
            - 모델은 입력 데이터를 바탕으로 특정 결과(예: 성공/실패)가 발생할 확률을 계산합니다.
            - 확률을 0과 1 사이의 값으로 변환하기 위해 시그모이드 함수(Sigmoid Function)를 사용합니다.
            - 예를 들어, 확률이 0.7 이상이면 "성공", 0.3 이하면 "실패"로 분류할 수 있어요.
    - 오즈 (odds)
        - 성공할 확률과 실패할 확률 비
        - odds = P/(1-P) = 성공확률/실패확률
    - 로짓(logit) 변환
        - 오즈에 자연로그(자연상수 e=2.71...가 밑)를 취해 선형 관계로 변환
        - loge(odds)=loge(P/(1-P))=알파+(베타*x)
            - P = 1/[1+e^{(-1)\*(알파+베타\*x)}], P는 0~1의 값
    - 시그모이드 함수
        - 로짓 함수의 역함수를 통해 0~1 확률 도출하는 함수
        - 독립변수 x가 n 증가시 확률이 e^n만큼 증가
- 오즈(Odds)
    - 어떤 일이 성공할 확률과 실패할 확률을 비교하는 개념이에요.
    - 예를 들어, 동전을 던져서 앞면이 나올 확률이 50%라면 실패할 확률도 50%죠.
        - 이걸 오즈로 표현하면 "1 대 1"이에요.
    - 그런데 만약 앞면이 나올 확률이 80%라면?
        - 실패할 확률이 20%니까 오즈는 "4 대 1"이 돼요. 즉, 성공할 가능성이 실패보다 4배 높다는 뜻이에요.
- 로짓(Logit) 변환
    - 이 오즈 값을 그대로 사용하면 계산이 복잡해요. 그래서 좀 더 다루기 쉽게 바꾸는 과정이 로짓 변환이에요.
    - 쉽게 말하면, 확률을 직선적인 숫자로 바꿔서 계산하기 쉽게 만드는 것이에요. 이렇게 하면 확률이 변할 때 좀 더 예측하기 쉬워져요.
- 시그모이드(Sigmoid) 함수
    - 이제 반대로, 변형된 숫자를 다시 확률 값(0~1)으로 바꿀 때 필요한 게 시그모이드 함수예요.
    - 예를 들면, 날씨에 따라 사람들이 우산을 들고 다닐 확률을 예측하고 싶다고 해볼게요.
        - 로짓 변환을 이용하면 날씨와 확률이 직선적인 관계를 가질 수 있지만, 결국 확률 값은 0과 1 사이에서 움직여야 하죠.
        - 이 시그모이드 함수가 그 역할을 합니다! 날씨가 흐려질수록 우산을 들 가능성이 점점 커지지만, 어느 순간부터는 거의 다 들고 다니니까 더 이상 급격히 증가하지 않는 것처럼요.
- 한 문장으로 정리하면!
    - 오즈는 확률을 비교하는 방식이고,
    - 로짓 변환은 확률을 다루기 쉽게 만드는 과정,
    - 그리고 시그모이드 함수는 계산된 값을 다시 0~1 사이의 확률로 바꿔주는 도구예요.

- KNN (K-Nearest Neighbors)
    - 거리 기반으로 이웃에 더 많은 데이터가 포함되어 있는 범주로 분류
    - 단순하고 효율적이며 훈련이 따로 필요없는 Lazy Model
    - K에 따라 결과가 달라짐
    - 세모데이터를 중심에 놓고 동그라미 치니까
        - 동그라미 안에 (세모 이웃한 곳에) x 4개 o 1개 있었다
        - 세모데이터는 x분류에 해당한다.
    - KNN (K-Nearest Neighbors) 분류
        - 이 방법은 새로운 데이터가 들어왔을 때, 가장 가까운 친구(K개)를 찾아보고 다수결로 결정하는 방식이에요.
        - 예를 들어, 당신 앞에 정체불명의 과일이 하나 있어요. 그런데 주변에 딸기 🍓 4개, 바나나 🍌 1개가 있다면? KNN 방식은 이 과일을 딸기로 분류해요, 왜냐하면 딸기가 더 많으니까요!
        - K값(친구의 수)을 늘리면 더 많은 주변 데이터를 참고하게 되고, 줄이면 가까운 소수의 데이터만 참고해요. 그래서 K값을 잘 설정하는 게 중요해요!
        - KNN은 가까운 친구들에게 물어보고 다수결로 결정!
- 나이브베이즈 분류
    - 나이브 베이즈(Naïve Bayes) 분류
        - 이 방법은 확률을 이용해서 가장 가능성이 높은 선택을 찾는 방식이에요.
        - 예를 들어, 날씨가 흐린 날 ☁️ 사람들이 우산을 들고 나오는 확률을 예측한다고 해볼게요.
            - "비가 올 때 사람들이 우산을 들고 나올 가능성"
            - "전체 날씨 중 비가 오는 날의 비율"
            - "사람들이 우산을 들고 나오는 전체 확률"
        - 이런 정보를 조합해서 오늘 흐린 날씨에 사람들이 우산을 들 가능성이 얼마나 되는지 계산하는 것이 나이브 베이즈 분류예요!
        - 쉽게 말하면, 기존의 확률을 이용해 가장 그럴듯한 선택을 예측하는 방법이죠.
        - 나이브 베이즈는 확률을 계산해서 가장 가능성이 높은 걸 선택!
    - 베이즈 정리
        - P(a|b)={P(b|a)*P(a)}/P(b)=P(a교b)/P(b)
            - P(a|b) - 사후확률 (Posterior Probability)
                - 어떤 조건이 주어졌을 때, 특정 사건이 발생할 확률이에요.
                - 예를 들어, 감기를 이미 앓고 있는 사람에게 "이 사람이 병원에 방문할 확률은?"을 묻는 것과 같아요. 즉, b(조건)이 이미 주어진 상태에서 a(사건)가 발생할 가능성을 보는 거죠!
            - P(b|a) - 우도 (Likelihood)
                - 특정 사건이 발생했을 때, 그 사건이 어떤 조건에서 나왔을 가능성이에요.
                - 예를 들어, 병원에 방문한 사람이 있다고 할 때, "이 사람이 감기일 확률은?"을 묻는 것과 같아요. 즉, a(병원 방문)이라는 사실을 알고 있을 때, b(감기)일 가능성을 찾는 거예요.
            - P(a) - 사전확률 (Prior Probability)
                - 아무 조건 없이 사건이 발생할 확률이에요.
                - 예를 들어, "평균적으로 감기에 걸릴 확률이 얼마나 될까?"를 묻는 것과 같아요. 즉, 감기에 걸린 사람을 따로 고려하지 않고 전체적인 확률을 보는 거죠.
            - P(b) - 주변우도 (Marginal Probability)
                - 주어진 사건이 전반적으로 발생할 확률이에요.
                - 예를 들어, 감기 여부를 고려하지 않고 "평균적으로 사람들이 병원에 방문할 확률이 얼마나 될까?"를 묻는 것과 같아요. 즉, 감기가 아니라 다른 이유로 병원에 방문한 경우도 포함하는 확률이에요.
            - P(a ∩ b) - a와 b가 동시에 발생할 확률
                - 둘 다 동시에 일어날 확률이에요.
                - 예를 들어, "감기에 걸린 사람이 병원에 방문할 확률은?"을 묻는 것과 같아요. 즉, 감기에 걸리면서 동시에 병원에 가는 경우를 계산하는 거죠.
        - 쉽게 말하면 "새로운 정보(b)를 알게 되었을 때, a가 일어날 확률을 업데이트하는 과정"이에요.
            - 예를 들어 "기침을 하는 사람이 감기에 걸렸을 확률"을 계산해본다고 해볼게요.
                - P(a) = 감기에 걸릴 확률 (전체 인구 중 감기 걸린 비율)
                - P(b|a) = 감기인 사람이 기침할 확률 (감기 걸린 사람 중 기침하는 비율)
                - P(b) = 기침하는 사람의 전체 확률 (감기 외에도 알레르기 등으로 기침할 수도 있죠)
                - 이 값을 베이즈 정리에 넣으면 "기침하는 사람이 감기일 확률"을 얻을 수 있어요!
                    - 사전확률 P(a): 감기에 걸릴 가능성 (일반적인 확률)
                    - 우도 P(ba): 감기 걸린 사람이 기침할 확률
                    - 주변우도 P(b): 모든 기침하는 사람의 확률 (감기뿐만 아니라 다른 원인도 포함)
                    - 사후확률 P(ab): 기침한 사람이 감기일 가능성
                    - P(a ∩ b): 감기에 걸리면서 병원에 방문할 확률 (둘 다 동시에 발생)
            - 즉, 새로운 정보가 들어올 때마다 확률을 업데이트하며 의사결정을 하는 것이 나이브 베이즈의 핵심이에요.

- 의사결정나무 (Decision Tree)
    - 노드 내 동질성은 커지고 노드 간 이질성은 커지도록 분리
    - 분할 방법
        - 분류(범주형)에서의 분할 방법
            - CHAID 알고리즘: 카이제곱 통계량
            - CART 알고리즘: 지니지수 활용: 1-시그마(P^2)
            - C4.5/C5.0 알고리즘: 엔트로피지수 활용: -시그마[P{loge(P)}]

            - CHAID 알고리즘: ANOVA, F-통계량
            - CART 알고리즘: 분산감소량
- 의사결정나무는 데이터를 체계적으로 분류하거나 예측하는 데 사용되는 방법입니다.
    - 핵심 개념은 "비슷한 데이터끼리 묶고, 서로 다른 데이터는 분리하는 것"이라고 보면 돼요.
    - 분류(범주형 데이터)의 분할 방법
        - 범주(카테고리)를 나누는 방법이에요.
        - 예를 들어, 고객이 "구매할 가능성이 높은지 낮은지"를 예측한다고 해볼게요.
            - CHAID 알고리즘  
                - 여러 범주(예: 나이대, 성별)에 대해 가장 관계가 깊은(즉, 통계적으로 연관성이 큰) 특징을 찾고 이를 기준으로 분할합니다.  
                - 예: 고객의 나이가 구매 여부와 가장 관계가 크다면, 나이대를 기준으로 그룹을 나누는 방식이에요.
            - CART 알고리즘  
                - 데이터를 나눌 때 “서로 다른 그룹끼리 최대한 차이가 나도록” 분리하는 방법이에요.  
                - 예: 고객 그룹을 "충성 고객"과 "일회성 구매 고객"으로 나누는 데 유용해요. 충성 고객일 확률을 높이는 방향으로 데이터를 나눕니다.
            - C4.5/C5.0 알고리즘
                - 정보를 가장 효율적으로 나누는 기준을 찾는 방식이에요.  
                - 예: “고객의 소득 수준이 구매 가능성을 가장 잘 설명하는 요소라면?” 소득 수준을 기준으로 그룹을 나누는 거죠.
    - 회귀(연속형 데이터)의 분할 방법
        - 숫자로 된 값을 예측할 때 쓰이는 방법이에요.
        - 예를 들어, “한 고객이 다음 달에 얼마를 소비할까?”를 예측한다고 가정해볼게요.
            - CHAID 알고리즘  
                - ANOVA 같은 통계 기법을 사용해 데이터를 그룹으로 나눠요.  
                - 예: 고객의 월급에 따라 소비 금액을 그룹화할 수 있어요. 월급이 높을수록 소비가 많은지 살펴보는 방식이죠.
            - CART 알고리즘  
                - 데이터를 나눌 때 "분산이 줄어드는 방향"으로 나눠요. 즉, 같은 그룹 내 데이터는 가능한 한 비슷한 값을 가지도록 만들어요.  
                - 예: 주식 시장에서 특정 주식의 가격 변동을 예측할 때, 비슷한 변동 패턴을 가진 데이터끼리 묶어서 분석할 수 있어요.

- 의사결정나무 이어서
    - 예시: 동전 던지기 지니지수와 엔트로피지수 계산
        - 앞면 확률 = 3/5
        - 뒷면 확률 = 2/5라고 할 때
        - 지니지수 = 1 -3/5제곱 -2/5제곱 = 12/25
        - 엔트로피지수 = -3/5*(로그e_3/5)-2/5*(로그e_2/5)
    - 지니지수와 엔트로피지수는 의사결정나무에서 분할 기준을 평가할 때 사용하는 척도입니다.
    - 쉽게 말해, "데이터가 얼마나 섞여 있는지"를 수치로 표현하는 방법이에요!
    - 지니지수 (Gini Index)
        - 데이터가 얼마나 순수한지(즉, 한 그룹 안의 데이터가 얼마나 비슷한지)를 측정하는 값이에요.
        - 값이 0에 가까울수록 같은 그룹 내 데이터가 동질적이고, 1에 가까울수록 이질적입니다.
        - 예를 들어, 동전 던지기에서 앞면이 나올 확률이 3/5, 뒷면이 나올 확률이 2/5일 때:
            - 지니지수 = 1 - (3/5)² - (2/5)² = 12/25
            - 즉, 완전히 순수한 그룹은 아니지만, 어느 정도 한쪽으로 치우친 상태라는 걸 알 수 있어요.
    - 엔트로피지수 (Entropy)
        - 데이터의 혼란도(불확실성)를 나타내는 값이에요.
        - 값이 0이면 완전히 순수한 데이터, 높을수록 다양한 요소가 섞여 있는 상태입니다.
        - 동전 던지기 예제에서:
            - 엔트로피지수 = - (3/5 × loge(3/5)) - (2/5 × loge(2/5))
            - 즉, 데이터가 얼마나 예측하기 어려운지를 나타내는 수치예요.
    - 차이점?
        - 지니지수는 계산이 더 단순하고 빠른 경향이 있습니다.
        - 엔트로피지수는 정보이론에서 나온 개념이라, 불확실성을 좀 더 정밀하게 측정해요.
    - 결론적으로, 의사결정나무는 더 순수한 그룹으로 데이터를 나누려면 어떤 기준을 적용해야 할까?를 평가할 때 지니지수와 엔트로피지수를 활용하는 거죠!
    -그룹 내 데이터란?
        - 의사결정나무는 데이터를 특정 기준을 이용해 여러 그룹으로 나눕니다.
        - 예를 들어, 학생들의 시험 성적을 분석한다고 할 때:
            - 그룹 A: 성적이 높은 학생
            - 그룹 B: 성적이 낮은 학생
        - 이런 식으로 데이터를 나누는 거죠. 이때, 같은 그룹 내 학생들은 성적이 비슷한 특성을 공유한다고 볼 수 있습니다.
    - 순수하다는 의미는?
        - 순수하다는 것은 그룹 안의 데이터들이 서로 비슷하다는 의미예요.
        - 예를 들어:
            - 그룹 A에는 모두 성적이 높은 학생들만 포함된다면 → 순수한 그룹
            - 그룹 A에 성적이 높은 학생도 있고 낮은 학생도 섞여 있다면 → 덜 순수한 그룹
        - 즉, 순수한 그룹을 만든다는 건 같은 특성을 가진 데이터끼리 묶어서, 서로 다른 데이터와 구분하는 것이에요.
    - 예시: 동전 던지기
        - 앞면이 나올 확률이 3/5, 뒷면이 나올 확률이 2/5인 경우:
            - 앞면만 포함된 그룹 (순수함)
            - 앞면과 뒷면이 섞인 그룹 (덜 순수함)
        - 결론적으로, 의사결정나무는 데이터를 최대한 순수한 그룹으로 나누려는 목표를 가지고 있어요!

- 의사결정나무 과적합 방지 방안
    - 정지규칙: 분리를 더 수행하지 않고 나무 성장 멈춤
    - 가지치기: 일부 가지 제거해 과대적합 방지
- 의사결정나무는 너무 깊어지면 과적합될 가능성이 높은데, 이를 방지하기 위한 방법이 정지규칙과 가지치기입니다.
    - 과적합(overfitting)은 머신러닝에서 모델이 훈련 데이터에 너무 지나치게 맞춰져서, 새로운 데이터에 대한 예측이 잘 안 되는 문제를 의미해요.
    - 정지규칙 (Stopping Rule)
        - 말 그대로 나무 성장을 멈추는 규칙이에요.
        - 데이터를 계속 나누다 보면 노드가 너무 많아지고, 너무 세분화되면 오히려 일반화 능력이 떨어질 수 있어요.
        - 그래서 특정 기준을 설정해서 "이제 그만 나누자!"라고 결정하는 거죠.
        - 예: 그룹을 더 나누려면 최소한 10개의 데이터가 있어야 한다는 규칙을 적용하면, 9개 이하라면 분할을 멈춰요.
            - 또 다른 방법은 트리의 최대 깊이를 제한하는 거예요. 예를 들어, 트리가 5층 이상 커지지 않도록 하면 과적합을 방지할 수 있죠.
    - 가지치기 (Pruning)
        - 이미 만들어진 의사결정나무에서 필요 없는 가지(분할된 노드)를 잘라내는 과정이에요.
        - 나무가 너무 복잡하면 특정 데이터에는 잘 맞지만, 새로운 데이터에 대한 예측력이 떨어질 수 있어요.
        - 가지치기는 불필요한 세분화를 줄이고, 더 단순한 구조로 만드는 방법이에요.
        - 예: 분할된 노드 중에서 예측력이 별로 도움이 안 되는 노드는 제거해서 트리를 더 깔끔하게 만듭니다.
            - 사후 가지치기(post-pruning): 트리를 완전히 만든 후에 가지를 잘라냄.
            - 사전 가지치기(pre-pruning): 트리를 만들면서 미리 가지를 최소화함.
    - 쉽게 말하면?
        - 정지규칙: 너무 깊이 분할하지 않도록 미리 제한을 두는 방법.
        - 가지치기: 이미 만들어진 트리에서 불필요한 부분을 제거해 더 일반화 능력을 높이는 방법.
        - 이렇게 하면 모델이 훈련 데이터에만 너무 의존하지 않고 새로운 데이터에 대한 예측 능력을 높일 수 있어요!
