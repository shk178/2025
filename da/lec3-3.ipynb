- 가설검정
    - 모집단의 특성에 대한 주장을 가설로 세우고 표본조사로 가설의 채택여부를 판정
    - 귀무가설(H0): 일반적으로 생각하는 가설 (차이가 없다)
    - 대립가설(H1): 귀무가설을 기각하는 가설, 증명하고자 하는 가설 (차이가 있다, 크다작다)
    - 유의수준(알파, 0.05=틀릴 확률 5%까지 허용): 귀무가설이 참일 때 기각하는 1종 오류를 범할 확률의 허용 한계
    - 유의확률(p-value): 귀무가설을 지지하는 정도를 나타내는 확률
    - 검정결과-실제
        - H0가 사실이라고 판정
            - 실제로 H0가 사실: 옳은 결정
            - 실제로 H0가 거짓: 1종 오류(알파)
        - H0가 거짓이라고 판정
            - 실제로 H0가 사실: 2종 오류(베타)
            - 실제로 H0가 거짓: 옳은 결정
    - 🍜 사례: 라면 제조사의 신제품 출시 실험
        - 한 라면 제조사가 새로운 라면이 기존 라면보다 더 맛있다고 주장한다고 해보자. 이 가설을 검증하기 위해 가설검정을 진행할 수 있어.
        - 🏁 가설 설정
            - 귀무가설(H₀): 새로운 라면과 기존 라면의 맛에는 차이가 없다. (즉, 새 라면이 더 맛있다고 할 수 없음)
            - 대립가설(H₁): 새로운 라면이 기존 라면보다 더 맛있다.
        - 📊 데이터 수집
            - 연구팀은 100명의 참가자를 모집해 블라인드 테스트를 진행한다.
            - 참가자는 랜덤하게 새 라면과 기존 라면을 먹고, 어느 쪽이 더 맛있는지 평가한다.
        - 🔍 결과 분석
            - 설문조사 결과, 새 라면이 기존 라면보다 맛있다고 답한 사람이 70명이었다.
            - 통계적으로 이 차이가 유의미한지 검정하기 위해 유의수준(α = 0.05)을 설정하고 p-value를 계산한다.
        - 🎯 검정결과
            - 만약 p-value가 0.03이라면? → 귀무가설(H₀)을 기각하고, 새 라면이 더 맛있다고 결론 내릴 수 있음. (차이가 유의미함)
            - 만약 p-value가 0.08이라면? → 귀무가설(H₀)을 기각하지 않음. 즉, 데이터만 봐서는 새 라면이 더 맛있다고 확신할 수 없음.
        - ⚠️ 오류의 가능성: 두 가지 오류가 발생할 수도 있어
            - 1종 오류(α 오류): 새 라면이 기존 라면보다 더 맛있지 않지만, 검정 결과에서는 더 맛있다고 결론 내린 경우.
                - 마치 효과 없는 건강식품을 보고 "우와, 건강에 좋네!"라고 믿어버리는 상황과 비슷해.
            - 2종 오류(β 오류): 새 라면이 실제로 더 맛있지만, 검정 결과에서는 차이가 없다고 판정한 경우.
                - 즉, 좋은 혁신 제품이 나왔는데도 아무도 알아보지 못한 상황이야.
        - 이렇게 가설검정을 통해 우리는 데이터를 기반으로 결정을 내리고, 오류의 가능성을 고려하는 과정을 거치게 돼!

- 가설 검정 문제 풀이 방법
    - 귀무가설/대립가설 설정
        - 차이가없다 혹은 동일하다 = 귀무가설
    - 양측/단측 검정 확인
        - 대립가설 값이 같지 않다 = 양측검정
        - 값이 크다/작다 = 단측검정
    - 일표본/이표본 확인
        - 하나의 모집단 = 일표본
        - 두개의 모집단 = 이표본
    - 귀무가설 기각/채택
        - p-value < 유의수준알파 = 귀무가설 기각
        - p-value > 유의수준알파 = 귀무가설 채택
    - t검정인 경우: 단일표본/대응표본/독립표본 확인
        - 모집단에 대한 평균검정 = 단일표본
        - 동일 모집단에 대한 평균비교 검정 = 대응표본
        - 서로 다른 모집단에 대한 평균비교 검정 = 독립표본
- 가설검정 문제를 푸는 과정을 쉽게 정리해볼게. 
    - 가설 설정: 먼저 가설을 세워야 해!  
        - 귀무가설(H₀): 차이가 없다, 동일하다 → 기본적으로 유지되는 가설  
        - 대립가설(H₁): 차이가 있다 → 우리가 증명하고 싶은 가설  
    - 검정 방법 결정
        - 양측검정: "차이가 있다" → 값이 크거나 작을 수 있음 (예: 평균이 다름)  
        - 단측검정: "값이 크다 또는 작다" → 특정 방향으로만 검정 (예: 새 제품이 더 나은지 확인)  
    - 표본 유형 확인
        - 일표본: 하나의 모집단에서 데이터 수집 → 한 집단만 분석  
        - 이표본: 두 개의 모집단 비교 → 서로 다른 집단의 차이 검정  
    - 결과 해석
        - p-value < α (유의수준) → 귀무가설 기각 → 차이가 있다!  
        - p-value > α → 귀무가설 채택 → 차이가 없다고 본다.  
    - t-검정 종류 결정
        - 단일표본 t-검정: 한 집단의 평균이 특정 값과 다른지 확인  
        - 대응표본 t-검정: 같은 모집단에서 두 상황을 비교 (예: 같은 사람의 전후 테스트 점수)  
        - 독립표본 t-검정: 두 다른 집단의 평균 비교 (예: 남성과 여성의 평균 키 비교)  
    - 이렇게 순서대로 접근하면 가설검정을 체계적으로 이해하고 문제를 풀 수 있어!

- 두 학교 수학 점수에 대한 t검정 예제
- `t.test(학교a, 학교b, conf.level=0.95)`
- `welch two sample t-test`
- `data: 학교a and 학교b`
- `t = -0.59758, df = 97.409, p-value = 0.5515`
- `alternative hypothesis = true difference in means is not equal to 0`
- `95 percent confidence interval:`
- `-10.293157      5.528998`
- `sample estimates:`
- `mean of x 61.91373`
- `mean of y 64.29581`
- 귀무가설/대립가설 설정
    - 차이가없다 혹은 동일하다: 귀무가설로 설정
        - 두 학교의 성적은 동일하다
    - 양측혹은단측검정 확인:
        - 대립가설의값이 같지않다 "양측검정"
    - 일표본혹은이표본 확인:
        - 두개의모집단 "이표본"
    - 귀무가설기각혹은채택
        - pvalue: 0.5515 유의수준 0.05 -> 귀무가설 "채택"
    - 단일/대응/독립표본 확인
        - 서로다른모집단에 대한 평균비교 검정 -> "독립표본"
- 이 예제는 두 학교의 수학 점수를 비교하는 독립표본 t-검정을 수행한 것이야. 간단히 해석해볼게.
    - 가설 설정
        - 귀무가설(H₀): 두 학교의 평균 점수가 동일하다. (차이가 없다)
        - 대립가설(H₁): 두 학교의 평균 점수가 다르다. (차이가 있다)
    - 검정 방법 선택
        - 두 모집단(학교A, 학교B)을 비교하는 것이므로 이표본 검정
        - "같지 않다"는 표현이 있으므로 양측검정을 수행  
        - 서로 다른 모집단이므로 독립표본 t-검정을 사용  
    - 결과 해석
        - t-value = -0.59758 → 두 학교 평균의 차이를 계산한 값  
        - p-value = 0.5515 → 귀무가설을 기각할지 결정하는 기준  
    - 결론
        - 보통 유의수준(α)을 0.05로 설정하는데, p-value가 0.5515로 훨씬 크다.  
        - 따라서 귀무가설을 기각하지 않음 → 즉, 두 학교의 평균 점수 차이가 통계적으로 유의미하지 않다.  
        - 95% 신뢰구간(-10.29 ~ 5.53)을 보면, 차이가 0을 포함하고 있어 의미 있는 차이가 있다고 볼 수 없음.  
    - 최종 해석
        - 이 검정 결과에 따르면 두 학교의 수학 성적은 유의미한 차이가 없다고 결론 내릴 수 있어!
        - 하지만 p-value가 크기 때문에, 더 많은 표본을 추가해서 검정을 다시 수행하면 더 정확한 결과를 얻을 수도 있어.

- 비모수 검정
    - 모집단에 대한 아무런 정보가 없어 관측자료가 특정분포를 따른다고 가정할 수 없을 때 검정
    - 두 관측 값의 순위나 차이로 검정
- 비모수 검정은 모집단에 대한 정보가 부족해서 특정한 분포를 가정할 수 없을 때 사용하는 검정 방법이야.
    - 핵심 개념 → 데이터를 직접 활용해 순위나 차이를 비교하는 방식!
    - 예제 상황 → "학생들의 시험 점수가 정규분포를 따르는지 모르지만, 두 그룹의 점수를 비교하고 싶어!"  
    - 대표적인 비모수 검정 방법:
        - 부호검정 → 두 집단의 중앙값 비교 (값이 증가했는지 여부만 확인)
        - 순위합검정/만휘트니 U검정 → 두 집단의 평균 순위 비교 (데이터 순서만 고려)  
        - 크러스칼-월리스 검정 → 세 개 이상의 집단을 비교하는 확장된 순위 검정  
        - 카이제곱검정 → 범주형 데이터(예: 설문조사 결과)를 비교하는 검정  
- 즉, 비모수 검정은 데이터의 분포를 몰라도 직접 비교할 수 있는 방법이라고 보면 돼!
