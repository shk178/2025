- 기초 확률 이론
    - 확률: 통계적 현상의 확실함 척도, 수학적확률/통계적확률 구분
    - 사건: 여러반복시행 통해 결과로 나타나는 표본공간의 부분집합이다.
    - 표본공간: 통계적실험에 의해 일어날 수 있는 모든 가능한 결과
        - 예: 동전 두개 던질 때 표본공간 s={(앞,앞),(앞,뒤),(뒤,앞),(뒤,뒤)}
    - 확률변수: 표본공간의 각원소에 해당하는 값(확률)을 대응하는 함수
        - 예: 확률변수 x가 어떤 집합의 키를 나타낼때/ 키가160~170일 확률 P(160<=X<=170)
    - 조건부확률: 특정사건b가 발생했을 때 a가 발생할확률
        - p(a|b)=p(a교b)/p(b) 백신을맞았을때 감기걸릴확률
    - 독립사건: a,b가 서로영향주지 않는사건 (p(a|b)=p(a))
        - p(a교b)=p(a)*p(b) 주사위a가3나올때 b도3나올확률
    - 배반사건: a,b가 서로동시에일어나지 않는사건
        - p(a교b)=공집합 동전던질때 앞뒷면동시나올확률
    - 베이즈정리: 두 확률변수의 사전확률과 사후확률 사이의 관계를 나타내는 정리
        - p(a|b)=p(b|a)*p(a)/p(b)
        - 베이즈 정리는 조건부 확률을 이용해서 유도할 수 있어.
            - 먼저, 어떤 두 사건 A와 B가 동시에 일어날 확률은 두 가지 방식으로 표현할 수 있어:
                - A가 일어난 후 B가 일어날 확률을 생각하면 p(b|a)
                    - A가 일어날 확률 × A가 일어난 후 B가 일어날 확률
                - B가 일어난 후 A가 일어날 확률을 생각하면 p(a|b)
                    - B가 일어날 확률 × B가 일어난 후 A가 일어날 확률
            - 이 두 값은 서로 같으니까,
            - (A가 일어난 후 B가 일어날 확률) × (A가 일어날 확률) = (B가 일어난 후 A가 일어날 확률) × (B가 일어날 확률)
            - 여기서 정리하면 A가 일어날 확률을 알고 있을 때 B가 일어날 확률을 구하는 공식이 나오고, 이게 바로 베이즈 정리야!

- 확률분포: 확률 변수의 개별 값들이 가지는 확률 값의 분포
    - 이산 확률분포: 값을 셀 수 있는 분표, 확률질량함수로 표현
        - 이산 균등분포
            - 모든 곳에서 값이 일정한 분표
            - 예: 주사위의 각 면이 나오는 확률 모두 동일
        - 베르노이 분포
            - 결과가 둘 중 하나로 나타나는 베르누이시행 분포
            - 예: 동전 던지기, 시험 합불
        - 이항분포
            - n번 베르누이 시행 중 k번 성공할 확률 분포
            - 예: 동전 20번 던져 앞면 나올 횟수
        - 기하분포
            - 성공확률 p인 베르누이시행에서 처음 성공시까지 시행횟수 분포
            - 예: 동전 던쳐 첨 앞면 나오기까지 던진 횟수
        - 음이항분포
            - 성공확률 p인 베르누이에서 성공 r번까지 반복시행 횟수 분포
            - 예: 동전 앞면 5번 나오기까지 던진 횟수
        - 초기하분포
            - N개 중 비복원추출 n번 추출시 원하는 결과 k번 나올 확률 분포
            - 예: 10개 구슬 중 4개의 구슬이 당첨 구슬이다. 이때 4번 뽑았을 때 당첨 구슬 2번 뽑을 확률은
                - 당첨 구슬(4개) 중에서 2개를 선택하는 경우의 수 → 4개의 당첨 구슬 중 2개를 뽑는 경우의 수는 6가지야.
                - 비당첨 구슬(6개) 중에서 2개를 선택하는 경우의 수 → 6개의 비당첨 구슬 중 2개를 뽑는 경우의 수는 15가지야.
                - 전체 구슬(10개) 중에서 4개를 선택하는 경우의 수 → 10개의 구슬 중 4개를 뽑는 경우의 수는 210가지야.
                - 이제 확률을 계산하면:
                    - (당첨 구슬에서 2개를 뽑는 경우의 수 × 비당첨 구슬에서 2개를 뽑는 경우의 수) ÷ 전체 구슬에서 4개를 뽑는 경우의 수
                    - (6 × 15) ÷ 210 = 90 ÷ 210 = 약 0.4286 (42.86%)
        - 다항분포
            - N번 시행에서 각 시행이 여러 개의 결과를 가질 수 있는 확률 분포
            - 예: 주사위 20번 던져 각 면이 나오는 횟수
        - 포아송분포
            - 단위 시간 내 발생할 수 있는 사건의 발생 횟수에 대한 분포
            - 예: 하루동안 출생자 수, 한시간동안 전화 걸려온 수
    - 연속 확률분포: 값을 셀 수 없는 분포, 확률밀도함수로 표현
        - 정규분포
            - 우리가 일상생활서 보는 확률변수의 평균분포를 근사한 분포 (Z검정 활용)
            - 예: 키분포, 시험성적분포
        - t분포
            - 정규분포와 유사하나 꼬리가 두껍고 긴 분포
            - 예: t검정 활용 - 표본이 30개보다 작은 집단에 대한 평균 검정
        - 카이제곱분포
            - 독립적인 정규분포를 따르는 변수들의 제곱합으로 구성된 분포
            - 예: 카이제곱검정 활용 - 두 집단 동질성 검정, 단일집단 모분산 검정
        - f분포
            - 두 개의 서로 다른 카이제곱분포의 비율
            - 예: 두 집단의 분산 동질성 검정

- 확률분포 기댓값
    - 확률분포 X의 f(x) 확률분포에 대한 기댓값(E(X))
        - 이산적 확률변수
            - E(X) = 시그마(x*f(x))
            - 예: 동전 3개 던질 때 확률변수 X(앞면F의 개수)의 기댓값은?
                - p(x=0)=bbb=1/8, p(x=1)=fbb,bfb,bbf=3/8, p(x=2)=ffb,fbf,bff=3/8, p(x=3)=fff=1/8
                - x값곱하기 확률을 더한다..1/8+3/8+6/8+3/8=1.5
        - 연속적 확률변수
            - E(X) = 인테그랄(x*f(x))
            - 예: 1~12 숫자 표시된 원형시계에서 확률변수 X(시계바늘이 가르키는시간)의 기댓값은?
                - 기댓값(E(X))은 확률변수 X의 평균적인 값을 의미해. 쉽게 말하면 "시계 바늘이 가리키는 시간의 평균적인 위치"를 찾는 거야.
                - 기댓값을 구할 때 연속적인 확률변수의 개념을 사용하면, 특정 구간에서 확률 밀도 함수(f(x))를 적분해서 평균적인 값을 계산하는 거야.
                - 이 문제에서는 시간(X)이 0(원형시계니까)~12 사이에서 균등한 확률로 분포하고, 확률 밀도 함수 f(x) 값이 1/12로 일정해.
                - 그래서 기댓값을 구할 때 x에다가 f(x)를 곱하고, 이를 0부터 12까지 적분한 후 계산하면
                - 확률밀도 함수가 x/12라면 적분할 함수는 x²/24, x=12니까 6이 나온다.

- 중심극한정리
    - 임의의 모집단으로부터 추출된 표본분포는 표본크기가 충분히 크면(30개이상) 정규분포
    - 모집단의 분포에 "상관없이" 표본"평균"분포가 정규분포를 이룬다. (n개짜리 표본을 m개 뽑아서 그 평균을 냈을 때)
        - 모집단의 평균u, 표준편차 시그마
        - 표본의 평균u, 표준편차 시그마/(루트n) // n은 표본 몇개뽑았는지
- 표본평균의 표본분포
    - 표본평균의 표본분포의 평균: E(바X)=u
    - 표본평균의 표본분포의 분산: V(바X)=(시그마제곱)/n //n은 표본크기
    - 표본평균의 표준화: Z=(바X-u)/(시그마/루트n)~N(0,1)
        - N: Normal Dist 정규 분포
        - 표본평균의 표준화는 간단히 말하면, 다양한 표본 평균값을 비교하기 쉽게 일정한 기준으로 변환하는 과정이야.
        - 예를 들어, 어떤 모집단에서 여러 개의 표본을 추출하면 각 표본마다 평균값이 조금씩 다를 거야. 하지만 이 표본평균을 그대로 쓰면 값의 크기가 다 다르기 때문에 비교하기 어려워. 그래서 표준화를 하면 모든 표본평균을 동일한 기준으로 변환해 쉽게 비교할 수 있어.
        - 표준화 과정에서 Z값을 계산하는데, 이는 표본평균이 전체 모집단의 평균(μ)에서 얼마나 떨어져 있는지를 측정하는 값이야. 이를 구할 때 표준편차를 반영해서 비교하는데, 표본 크기가 클수록 표본평균이 모집단 평균에 더 가까워지는 특성을 반영하기 위해 표본 크기의 제곱근(√n)을 사용해 보정하지!
        - 결국 Z값은 특정 표본평균이 전체 분포에서 얼마나 일반적인 값인지, 즉 평균에서 얼마나 멀리 떨어져 있는지를 알려주는 기준이라고 생각하면 쉬워. 이 값은 표준정규분포를 따르기 때문에, 여러 다른 표본에서도 동일한 기준으로 비교할 수 있어.

- 점추정
    - 모집단이 특정한 값으로 추정하며, 추정량(estimator)으로 모수를 측정
    - 추정량의 조건
        - 불편성 (Unbiasedness): 추정량의 기댓값이 실제 모수와 같음 (편향이 0이 되는 경우)
        - 효율성 (Efficiency): 여러 추정량 중 분산이 작은 것이 더 효율적인 추정량
        - 일치성 (Consistency): 표본 크기가 증가할수록 추정량이 모수에 가까워짐
        - 충족성 (Sufficiency): 추정량이 모집단의 정보를 최대한 반영
    - 대표적인 추정량
        - 모집단의 평균 u -> 표본평균 바X=(1/n)*(시그마X)
        - 모집단의 분산 시그마제곱 -> 표본분산 s제곱=(1/(n-1))*(시그마{(X-바X)의제곱})
            - 모집단 큰원, 표본집단 작은원이라고 했을 때 분산이 표본집단이 작을 수밖에 없다 (원지름)
            - 그래서 평균을 n으로 나누고 분산은 그보다 작은 n-1로 나눈다.
    - 점추정을 쉽게 설명하면, 모집단의 어떤 값(평균이나 분산 등)을 알고 싶을 때, 우리가 가진 표본 데이터를 바탕으로 하나의 숫자로 그 값을 추정하는 과정이야.
    - 예를 들어, 어떤 도시의 평균 키를 알고 싶다고 해보자. 당연히 모든 사람의 키를 다 측정할 수 없으니까, 일부 사람들을 랜덤하게 선택해서(표본) 그들의 평균 키를 계산하고, 이를 도시 전체의 평균 키에 대한 추정값으로 사용할 수 있어. 이처럼 모집단의 실제 값(모수)을 하나의 숫자로 추정하는 것이 바로 점추정이야.
    - 하지만 추정값이 믿을 만해야 하겠지? 그래서 좋은 추정량(Estimator)은 몇 가지 조건을 만족해야 해:
        - 불편성: 너무 치우치지 않고, 실제 값과 가깝게 추정해야 해.
        - 효율성: 여러 가지 방법 중 가장 정확하고 흔들림(분산)이 적어야 해.
        - 일치성: 표본 크기가 커질수록, 점점 더 실제 값에 가까워져야 해.
        - 충족성: 모집단의 정보를 최대한 잘 반영해야 해.
    - 대표적인 예로 표본평균을 쓰면 모집단의 평균을 추정할 수 있고, 표본분산을 사용하면 모집단의 분산을 추정할 수 있어.
    표본평균과 표본분산을 사용해서 모평균과 모분산을 추정하지만, 정확히 같지는 않아.
        - 모평균(μ)은 모집단 전체의 평균이고, 표본평균(𝑋̄)은 일부 표본을 통해 계산한 값이야. 표본을 통해 추정할 때, 표본평균이 모평균과 거의 비슷해지지만, 완전히 같다고 단정할 수는 없어. 그렇기 때문에 표본평균을 모평균의 추정량으로 사용한다고 표현해.
        - 모분산(σ²)은 모집단 전체의 분산이고, 표본분산(s²)은 표본을 이용해 계산한 분산이야. 여기서 중요한 차이점이 하나 있어! 표본분산을 구할 때 분모가 (n-1)로 나누는 이유는 표본 데이터를 사용하면 모집단의 실제 분산보다 낮게 추정될 가능성이 있기 때문이야. 그래서 조금 더 정확한 값을 얻기 위해 자유도를 반영해서 보정하는 거야.
        - 표준편차도 비슷한 개념인데, 모표준편차(σ)는 모집단의 표준편차이고, 표본표준편차(s)는 표본으로 계산한 표준편차야. 표준편차는 분산의 제곱근이기 때문에, 표본분산과 마찬가지로 표본표준편차도 (n-1)로 나눈 값을 사용해.
    - 결론적으로, 모평균과 모분산을 직접 알 수 없기 때문에, 표본을 이용해서 추정하는 거야! 하지만 표본이 모집단을 정확히 대표하지 못할 수도 있어서, 통계적으로 보정을 한다는 점을 기억하면 좋아

- 구간추정(신뢰구간)
    - 모집단이 특정한 구간으로 추정(95%, 99%를 가장 많이 사용)
    - 모집단의 분산(시그마제곱)을 알고 있는 경우
        - 신뢰수준 95%: u >=<= 바X+-(1.960)(시그마/루트n)
        - 신뢰수준 99%: u >=<= 바X+-(2.576)(시그마/루트n)
    - 모집단의 분산을 모르는 경우
        - 자유도가 n-1인 t분포로 신뢰구간 추정
            - 수학적으로 n개 표출하면 자유도가 n-1이 된다. (분산 작아지는 것)
            - 모분산 모르니까 시그마대신 표본표준편차s 쓴다.
            - 신뢰수준 95%: u >=<= 바X+-(t95%상수)(표본표준편차s/루트n)
            - 신뢰수준 99%: u >=<= 바X+-(t99%상수)(표본표준편차s/루트n)
