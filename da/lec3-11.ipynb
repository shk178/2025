- 분류모델 평가지표
    - 오분류표 (혼동행렬, Confusion Matrix)
        - 영화 추천 시스템이 사용자가 좋아할 만한 영화인지 아닌지 예측한다고 가정해 볼게요.  
        - 이때, 실제 사용자 반응(좋아함/싫어함)과 추천 모델의 예측(추천함/추천 안 함)을 비교할 수 있어요.
    - FP (잘못된 추천)와 FN (추천 실패) 중 더 위험한 것은?
        - FP가 많으면? → 사용자는 계속 관심 없는 영화를 추천받아 불편할 수 있어요!  
        - FN이 많으면? → 사용자가 좋아할 만한 영화를 추천받지 못해서 좋은 경험을 놓칠 수 있어요!

    | 실제 반응 | 추천 예측 | 설명 |
    |:-:|:-:|:-:|
    | True Positive (TP) 포지티브가 맞다 | ✅ 추천함 | 추천했더니 정말 좋아함 (완벽한 추천!) |
    | False Positive (FP) 포지티브 틀림(네거티브다) | ❌ 추천함 | 추천했는데 싫어함 (불필요한 추천) |
    | True Negative (TN) 네거티브가 맞다 | ✅ 추천 안 함 | 추천 안 했고, 실제로도 안 좋아함 (잘 걸러냄) |
    | False Negative (FN) 네거티브 틀림(포지티브다) | ❌ 추천 안 함 | 추천 안 했는데, 사실은 좋아할 영화였음 (좋은 영화 놓침) |

    - 정밀도 (Precision)
        - 정밀도 = TP / (TP + FP)  
        - 예: "추천한 10편 중 8편이 실제로 사용자가 만족했다!"  
        - FP가 많으면? → 잘못 추천한 영화가 많아져 정밀도가 낮아짐!  
    - 재현율 (Recall)
        - 재현율 = TP / (TP + FN)  
        - 예: "사용자가 실제로 좋아한 20편 중 15편을 추천했다!"  
        - FN이 많으면? → 사용자가 좋아할 영화를 놓칠 확률이 높아짐!  
    - 특이도 (Specificity)
        - 특이도 = TN / (TN + FP)  
        - 예: "사용자가 싫어하는 영화는 잘 걸러냈다!"  
        - FP가 많으면? → 추천 모델이 관심 없는 영화를 추천할 가능성이 높아짐!  
    - 정확도 (Accuracy)
        - 정확도 = (TP + TN) / 전체 데이터  
        - 예: "전체 추천 중 80%가 적절했다!"  
        - FP와 FN이 많으면? → 정확도가 낮아짐!  
    - FP Rate (거짓 경보율)
        - FP Rate = FP / (FP + TN)  
        - 예: "모델이 추천했는데, 사용자는 관심 없는 경우가 많다면 낮출 필요가 있음!"  
        - FP가 많으면? → 불필요한 추천이 많아져 사용자 불만 증가!  
    - F-1 Score (정밀도와 재현율 균형)
        - F-1 Score = 2 × (Precision × Recall) / (Precision + Recall)  
        - 예: "정확한 추천도 중요하지만, 많이 추천해야 할 수도 있다!"  
        - FP 또는 FN이 많으면? → 정밀도와 재현율 균형이 깨짐!  
    - F-베타 Score (정밀도 vs 재현율 중 더 중요한 쪽 강조)
        - F-베타 Score = (1 + β²) × (Precision × Recall) / (β² × Precision + Recall)  
        - 예: "사용자가 좋아할 영화를 놓치지 않는 게 가장 중요하다면 재현율을 높이고 싶다!"  
        - β 값이 크면? → 재현율(Recall)이 더 중요하게 고려됨!

    | 지표 | 예시 | 공식 | 영향을 받는 요소 | 추가 설명 |
    |:-:|:-:|:-:|:-:|:-:|
    | 정밀도 | 추천한 영화 중 실제로 좋아한 비율 | TP / (TP+FP) | FP 많으면 낮아짐 |정밀도-재현율은 Trade-off 관계|
    | 재현율 | 실제로 좋아한 영화 중 추천된 비율 | TP / (TP+FN) | FN 많으면 낮아짐 | =민감도=TP RATE=HIT RATE |
    | 특이도 | 싫어하는 영화 안 추천한 비율 | TN / (TN+FP) | FP 많으면 낮아짐 ||
    | 정확도 | 전체에서 맞춘 비율 | (TP+TN) / 전체 | FP, FN 많으면 낮아짐 ||
    | FP Rate | 불필요한 추천이 얼마나 많았는지 | FP / (FP+TN) | FP 많으면 높아짐 ||
    | F-1 Score | 정확한 추천 vs 많이 추천하는 것 사이 균형 | 2×(Precision×Recall) / (Precision+Recall) | FP 또는 FN 많으면 낮아짐 | 정밀도와 재현율의 조화평균 |
    | F-베타 Score | 정밀도와 재현율 중 더 중요한 쪽을 강조 | (1+β²)×(Precision×Recall) / (β²×Precision+Recall) | β 값에 따라 정밀도 or 재현율 강조 | 베타>1: 재현율비중큼, <1: 정밀도비중큼, =1: F-1 Score와 동일|

- 분류모델 평가지표 이어서
    - ROC 커브(Receiver Operating Characteristic Curve)
        - 가로축을 "1-특이도(FPR)", 세로축을 "민감도(TPR)"로 두어 시각화한 그래프
        - 그래프 면적(AUC)은 0.5~1사이이며, 1에 가까울수록 모델 성능 좋다고 평가
    - 이익도표 (Lift chart)
        - 임의로 나눈 각 등급별로 반응검출율, 반응률, 리프트 등의 정보를 산출해 나타내는 도표
        - 향상도 곡선: 이익도표를 시각화한 곡선
- ROC 커브는 분류 모델의 성능을 평가하는 그래프입니다.  
    - 모든 가능한 임계값(Threshold)을 변화시키면서 특이도(FPR)와 민감도(TPR)를 계산하여 그래프를 그립니다.
    - ROC 커브의 축
        - X축: 1-특이도 (False Positive Rate, FPR)  
            → 모델이 음성을 양성으로 잘못 분류한 비율 (오탐 확률)
        - Y축: 민감도 (True Positive Rate, TPR)  
            → 실제 양성 중에서 모델이 정확히 양성으로 분류한 비율 (검출률)
    - ROC 커브의 형태
        - AUC (Area Under the Curve)는 ROC 곡선 아래 면적을 뜻하며, 모델의 성능을 측정하는 지표입니다.
        - AUC 값이 1에 가까울수록 모델 성능이 좋고, 0.5에 가까울수록 랜덤 추측과 비슷합니다.
    -특수한 경우
        - AUC = 0.5 (랜덤 모델)
            - ROC 곡선이 대각선(45도 직선) 형태로 나타납니다.  
            - 즉, 랜덤으로 분류하는 것과 같은 수준이므로 모델 성능이 낮습니다.
            - 모델이 임의로 예측하는 경우 → 양성과 음성 구분이 불가능
                - 이때 ROC 곡선은 대각선 형태를 가짐
                - 즉, y = x와 같은 직선 그래프가 됩니다.
        - AUC = 1 (완벽한 모델)
            - ROC 곡선이 좌측 상단을 지나서 100% 민감도에서 수평으로 그려지는 형태입니다.  
            - 즉, 모든 양성 사례를 완벽하게 맞추면서 음성 사례는 전혀 오탐하지 않는 모델이 됩니다.
            - FPR = 0일 때, TPR = 1 (그래프의 시작점: 좌상단)
                - 이후 TPR은 1을 유지하면서 FPR이 증가하지 않음 → 수평으로 유지
                - 즉, y = 1 형태의 직선이 됩니다.
        - AUC < 0.5 (오류 모델)
            - ROC 곡선이 대각선 아래에 위치합니다.  
            - 즉, 모델이 양성과 음성을 반대로 예측하는 경우입니다.
            - 모델이 양성과 음성을 반대로 예측하면 FPR이 증가할 때 TPR이 감소
                - 즉, ROC 곡선은 y = 1-x 형태의 우하향 대각선이 됩니다.
                - 이런 경우 모델 출력을 뒤집어서(양성을 음성, 음성을 양성) 다시 평가하면 더 좋은 성능을 얻을 수도 있습니다.
- 이익도표는 마케팅, 금융, 고객 분석 등에서 모델의 성능을 평가하는 그래프입니다.  
    - 특정 기준(예: 예측 점수 등)을 기준으로 데이터를 여러 그룹으로 나누어 성능을 측정합니다.
    - 이익도표의 주요 지표
        - 반응검출율 (Cumulative Gain)  
            → 특정 그룹에서 모델이 올바르게 양성을 검출한 비율
        - 반응률 (Response Rate)  
            → 특정 그룹에서 실제 양성 비율
        - 리프트 (Lift)  
            → 특정 그룹에서 반응률이 전체 평균보다 몇 배 높은지를 나타냅니다.
    - 이익도표의 형태
        - 리프트 값 = 1 (무작위 모델)
            - 모든 그룹의 반응률이 전체 평균과 동일한 경우입니다.
            - 그래프는 수평 직선으로 그려집니다.
        - 리프트 값 > 1 (좋은 모델)
            - 높은 점수 그룹에서 반응률이 평균보다 높으면 리프트 곡선이 가파르게 상승합니다.
            - 즉, 모델이 높은 점수 그룹에서 양성 사례를 잘 찾고 있다는 뜻입니다.
        - 리프트 값 < 1 (잘못된 모델)  
            - 모델이 높은 점수 그룹에서도 반응률이 낮으면 곡선이 평균보다 아래로 위치합니다.  
            - 즉, 모델이 양성 사례를 효과적으로 찾지 못하고 있다는 의미입니다.

`군집분석`
- 군집분석
    - 비지도 학습으로 데이터들 간 거리나 유사성을 기준으로 군집을 나누는 방식
    - 거리측도
        - 연속형변수 (1,2) (4,4)
            - 유클리디안거리: 두점사이 직선거리
                - 루트(3제곱+2제곱)=루트13
            - 맨하튼거리: 변수들 차이의 단순 합
                - |4-1|+|4-2|=5 (x끼리 y끼리)
            - 체비셰프 거리: 변수 거리차 중 최댓값
                - max(|4-1|, |4-2|)=3
            - 표준화 거리: 유클리디안 거리를 표준편차로 나눔
            - 민코우스키 거리: 유클리드, 맨하튼 거리 일반화한 거리
            - 마할라노비스 거리: 표준화 거리에서 변수의 상관성 고려
        - 범주형 변수
            - 자카드 유사도: 합집합과 교집합의 비율
            - 코사인 유사도: 코사인 각도 활용
- 계층적 군집분석
    - 거리측정 방법
        - 최단연결법(단일연결법): 군집간 가장 가까운 데이터
        - 최장연결법(완전연결법): 군집간 가장 먼 데이터
        - 평균연결법: 군집의 모든 데이터들의 평균
        - 중심연결법: 두 군집의 중심
        - 와드연결법: 두 군집의 편차 제곱합이 최소가 되는 위치
    - 덴드로그램
        - 계층적 군집화를 시각적으로 나타내는 Tree모양의 그래프
- 덴드로그램에서 거리란?
    - 덴드로그램은 군집이 형성되는 과정을 트리 형태로 보여주는 그래프입니다.
    - 이때 거리는 두 데이터 또는 군집 사이의 유사성을 나타내는 값이에요.
    - 예를 들어 거리 15에서 나누면 3개의 클러스터가 만들어지고,
    - 거리 25에서 나누면 2개의 클러스터가 형성된다는 것은,
        - 군집을 나누는 기준이 15일 때는 더 작은 그룹으로 분리되고, 25일 때는 더 큰 그룹으로 묶인다는 뜻이에요.
- 거리 측정 방식에 따른 차이
    - 거리 측정 방식이 달라지면 덴드로그램의 구조도 변할 수 있어요.
    - 예를 들어, 같은 데이터라도 거리 계산 방법을 다르게 하면 군집 형성 시점이 달라질 수 있습니다.
    - 최단연결법 (Single Linkage) → 가장 가까운 두 점 사이의 거리
        - 데이터 간 최소 거리를 기준으로 군집을 형성.
        - 덴드로그램이 길게 늘어지는 형태가 될 가능성이 큼.
    - 최장연결법 (Complete Linkage) → 가장 먼 두 점 사이의 거리
        - 군집 내에서 최대 거리를 기준으로 묶음.
        - 덴드로그램의 군집이 좀 더 밀집된 형태로 나올 가능성이 큼.
    - 평균연결법 (Average Linkage) → 모든 데이터의 평균 거리
        - 군집 전체의 평균적 유사성을 반영.
        - 덴드로그램이 중간 형태로 그려질 가능성이 많음.
    - 와드연결법 (Ward Linkage) → 군집 내 분산이 최소가 되도록 묶음
        - 군집 내 데이터의 변동성을 최소화하는 방식
        - 균형 잡힌 덴드로그램이 나오게 됨.
- 덴드로그램에서 거리는 군집화 방식에 따라 다르게 계산됨.
- 거리 15에서 나누면 더 세분화된 군집, 거리 25에서 나누면 큰 군집이 형성됨.
- 최단연결법, 최장연결법, 평균연결법 등 거리 측정 방식에 따라 덴드로그램 모양도 달라짐.

- 군집을 나누는 방법 (거리 측정)
    - 군집을 나누려면 데이터 간의 거리를 측정해야 해요.

        → "가장 멀리 떨어진 친구와 가장 가까운 친구를 비교해서 그룹을 나누자!"  
        → 차이값 중 가장 큰 거리를 기준으로 군집을 만듭니다.  
    - 범주형 거리 예시 (자카드 유사도)  
        - "우리가 좋아하는 영화 리스트를 비교해서 비슷한 취향끼리 묶자!"  
        - 서로 본 영화 개수를 비교해서 군집을 만듭니다.  
- 계층적 군집분석 (연결법 예시)
    - 군집을 묶는 방법도 여러 가지가 있어요! 친구들끼리 그룹을 만들 때를 예로 들어볼게요.  
    - 최단연결법 (단일연결법)  
        → "가장 가까운 친구 둘부터 시작해서 점점 그룹을 키우자!"  
        → 예: A와 B가 서로 제일 친하니까 먼저 묶고, 그 다음 C를 추가.
    - 최장연결법 (완전연결법)  
        → "우리가 가장 멀리 떨어진 친구를 기준으로 그룹을 정하자!"  
        → 예: 그룹 내 가장 멀리 있는 사람을 고려해서 군집을 결정.
    - 평균연결법  
        → "우리 그룹의 평균 거리 기준으로 묶자!"  
        → 예: A, B, C의 평균 거리가 가장 가까운 그룹을 형성.
    - 중심연결법  
        → "그룹의 중심 위치를 계산해서 그룹을 묶자!"  
        → 예: 학급에서 중간 성적을 가진 학생들끼리 그룹을 만들기.
    - 와드연결법  
        → "각 그룹의 성격이 최대한 비슷하게 유지되도록 묶자!"  
        → 예: 비슷한 학습 성향을 가진 친구들끼리 그룹을 만든다고 생각해보세요!
- 덴드로그램 예시
    - 덴드로그램은 군집이 어떻게 형성되는지 보여주는 트리 구조 그래프예요.  
    - 예를 들어, 가족 관계를 표현하는 족보처럼 데이터가 점점 묶여가는 과정을 시각화하는 방식이에요.
    - 친구 관계를 덴드로그램으로 보면?  
        → 가장 가까운 친구들끼리 먼저 묶이고, 점점 그룹이 커지면서 큰 집단이 형성되는 과정을 볼 수 있어요!
- 군집분석은 비슷한 데이터를 그룹으로 묶는 방법이에요.  
- 거리 측정 방식에 따라 유클리디안 거리, 맨하튼 거리 등 여러 방법이 있음.  
- 연결법을 이용해 단계적으로 그룹을 키워나가는 방식(계층적 군집)이 존재함.  
- 덴드로그램을 이용하면 데이터가 그룹으로 묶이는 과정을 한눈에 볼 수 있음.
