> 다층회귀분석(Hierarchical Regression Analysis)은 여러 단계에 걸쳐 독립변수를 추가하여 분석하는 방법으로, 변수들이 결과 변수에 미치는 영향을 단계적으로 검토하는 과정이 포함됩니다. 이는 변수들의 상대적인 기여도를 평가하거나 통제변수를 추가하여 분석할 때 활용됩니다.

`하나 이상의 독립변수가 사용된 선형회귀 분석이며, 여러 독립변수를 +로 연결해 나열해 분석하는 기법을 무엇이라 하는가`
#### 다중선형회귀분석
> 다층선형회귀분석(Hierarchical Linear Regression)은 존재하지만, 일반적인 다층회귀(Hierarchical Regression)와 약간 다릅니다.
다층회귀분석은 변수들을 단계적으로 추가해가면서 모델이 개선되는지를 평가하는 기법으로, 변수가 추가됨에 따라 설명력이 어떻게 변하는지를 분석하는 데 중점을 둡니다. 반면, 다층선형회귀분석은 다중회귀를 계층적으로 적용하는 방식으로, 예를 들어 데이터가 그룹화되어 있어 집단 간 차이를 고려할 때 사용될 수 있습니다.

`최적회귀 방정식의 선택: 설명변수의 선택->가능한 범위 내에서 적은 수의 설명변수 포함`
- `#형식: stem(lm(종속변수~설명변수, 데이터세트))`
- `단계적 변수 선택에 3가지 방법 있다.`
    - `첫 번째 방법 (1)`
#### 전진선택법
        - `step(lm(Pemax ~ 1, Bio), scope=list(lower=~1, upper=나이+키+체중), direction="forward")`
            > - step() 함수를 이용한 단계적 회귀(stepwise regression)를 수행하는 과정이에요. 간단히 설명하면:
            > - lm(Pemax ~ 1, Bio): Pemax를 종속 변수로 하는 선형 회귀 모델을 만들되, 처음에는 상수항(Intercept)만 포함한 모델을 설정.
            > - scope=list(lower=~1, upper=나이+키+체중: 모델이 선택할 수 있는 변수들의 범위를 지정.
            > - lower=~1: 가장 단순한 모델(상수항만 포함).
            > - upper=나이+키+체중: 포함 가능한 모든 독립 변수.
            > - 즉, 이 코드는 Pemax를 예측하기 위해 Bio 데이터셋에서 변수들을 선택하는 과정에서 단계적 변수 선택(stepwise selection)을 수행하는 것이죠.
        - `절편만 있는 상수 모형으로부터 시작해 중요하다고 생각되는 설명변수부터 차례로 모형에 추가`
        - `(2)으로 가장 설명을 잘 하는 변수를 고려해 그 변수가 유의하면 추가, 그렇지 않으면 추가 멈춤`
#### 제곱합의 기준
    - `두번째 방법 (3)`

    - `상관계수: 두 확률변수 X,Y의 공분산을 각 확률변수의 표준편차의 곱으로 나눈 값`

#### 등간 척도

#### 서열척도
- `상관계수 계산할 두 데이터의 실제 값 대신 두 값의 순위rank를 사용해 상관계수 계산하는 방식, 값의 범위: [-1, 1]`
- `예: 국어점수-영어점수 간 상관관계: 피어슨상관계수`
- `예: 국어석차-영어석차 간 상관관계: 스피어만상관계수`
> 1. 이산형 데이터 (Discrete Data)
> - 특징: 연속적인 값이 아니라 특정 값만 가질 수 있는 데이터입니다.
> - 예시: 학생 수(10명, 11명처럼 정수 값만 가능), 출생년도(1990년, 1991년 등)
> - 상관계수 계산 방법: 이산형 데이터는 값들이 특정 범위를 가지므로 일반적으로 카이제곱 검정이나 다른 비모수적 방법을 사용해 분석하는 경우가 많습니다.
> 2. 순서형 데이터 (Ordinal Data)
> - 특징: 값의 크기나 순서가 의미가 있지만, 값들 간의 간격이 일정하지 않을 수도 있는 데이터입니다.
> - 예시: 시험 석차(1등, 2등, 3등), 만족도 조사(매우 만족, 만족, 보통, 불만족)
> - 상관계수 계산 방법: 피어슨 상관계수보다 스피어만 상관계수를 많이 사용합니다.
> - **스피어만 상관계수(Spearman Correlation Coefficient)**는 값의 크기 자체가 아니라 순위를 기반으로 계산합니다.
> - 예를 들어, 국어 점수와 영어 점수 간의 관계는 피어슨 상관계수로 계산하지만, 국어 성적 순위와 영어 성적 순위의 관계는 스피어만 상관계수를 사용해 분석하는 것이 더 적절합니다.
