    - 스피어만 상관분석: 서열 척도, 순서형 변수, 선형/비선형적 관계 나타냄

- 피어슨 상관계수
    - 가: +1, 나: +0.8, 다: -1, 라: -0.8
        - +1: 두 변수는 완벽한 양의 선형 관계를 가짐 (X가 증가하면 Y도 완벽하게 증가)
        - +0.8: 두 변수는 강한 양의 선형 관계를 가짐 (X가 증가하면 Y도 증가하지만 완벽한 선형은 아님)
        - -1: 두 변수는 완벽한 음의 선형 관계를 가짐 (X가 증가하면 Y는 완벽하게 감소)
        - -0.8: 두 변수는 강한 음의 선형 관계를 가짐 (X가 증가하면 Y도 감소하지만 완벽한 선형은 아님)
- 스피어만 상관계수: XY가 선형관계가 아니어도 +1 혹은 -1 될 수 있다.
    - 가: +1, 나: +1, 다: -1, 라: -1
        - +1: 두 변수의 순위(rank)가 완벽하게 같은 방향으로 움직임 (순위가 높으면 상대 변수도 순위가 높음)
        - +1 (나의 경우): 피어슨 계수는 0.8이었지만, 순위 기준으로 보면 완벽한 단조 증가 관계로 볼 수 있음
        - -1: 두 변수의 순위가 완벽하게 반대 방향으로 움직임 (순위가 높을수록 상대 변수의 순위는 낮아짐)
        - -1 (라의 경우): 피어슨 계수가 -0.8로 완전한 선형 관계는 아니었지만, 순위 기준에서는 완벽하게 반대 방향
- 피어슨 계수는 두 변수 간의 선형 관계(linear relationship)를 측정하고
    - 선형 관계 (Linear Relationship)
        - 숫자의 크기가 직접적으로 증가 또는 감소하는 패턴을 의미
        - 예시: 키와 몸무게
            - 일반적으로 키가 클수록 몸무게도 증가하는 경향이 있음 (완벽한 선형은 아님, 사람마다 다름)
            - 키가 160cm에서 170cm로 증가할 때 평균적으로 몸무게도 증가하지만 반드시 동일한 증가폭이 아닐 수도 있음
            - 이 경우 피어슨 상관계수를 사용하여 "얼마나 강한 선형 관계를 가지는지" 분석 가능
- 스피어만 계수는 변수 간 순위 관계(rank correlation)를 봅니다.
    - 순위 관계 (Rank Relationship)
        - 숫자가 아니라 순서가 같은 방향 또는 반대 방향으로 움직이는지를 의미
        - 예시: 학생들의 성적 순위와 운동 능력 순위
            - 시험 성적 1등인 학생이 운동 능력도 1등이면 강한 양의 순위 관계 (+1)
            - 시험 성적이 높아질수록 운동 능력이 낮아진다면 강한 음의 순위 관계 (-1)
            - 그러나 성적 점수 자체가 아니라 순위(Rank)를 기준으로 평가한다는 점에서 피어슨과 다름
            - 순위를 기준으로 하기에 스피어만 상관계수를 사용

- 주성분 분석 (PCA)
    - 상관성 높은 변수들의 선형 결합으로 차원을 축소해 새로운 변수 생성
    - 자료의 분산이 가장 큰 축이 첫 번째 주성분 (고유값 고려)
    - 70~90%의 설명력을 갖는 수를 결정
- 주성분 분석(PCA, Principal Component Analysis)는 데이터의 복잡성을 줄이면서 중요한 정보를 유지하는 기법입니다.
    - 차원 축소(Dimensionality Reduction)
        - 여러 개의 변수(예: 학생들의 성적, 키, 몸무게, 운동 능력)가 있다면, PCA는 이 변수들을 조합해 새로운 축(주성분)을 만듭니다.
        - 이 과정에서 상관성이 높은 변수들끼리 묶어 새로운 변수(주성분)를 생성합니다.
    - 첫 번째 주성분의 의미
        - 데이터에서 가장 큰 분산(정보량이 가장 많음)을 가지는 방향을 첫 번째 주성분(PC1)으로 설정합니다.
        - 주성분들은 서로 독립적(uncorrelated)이며, 고유값(Eigenvalue)을 사용해 중요도를 평가합니다.
    - 설명력(Variance Explained)
        - PCA를 수행하면, 주성분들이 원래 데이터의 정보를 얼마나 설명할 수 있는지 계산합니다.
        - 일반적으로 70~90% 이상의 설명력을 가지는 주성분만 선택하여 데이터의 핵심 정보만 유지합니다.
    - 예제: 학생 성적과 학습 패턴 분석
        - 학생들의 데이터(시험 점수, 공부 시간, 출석 횟수 등)가 있다고 가정해보죠.
        - 시험 점수와 공부 시간은 강한 양의 상관 관계를 가질 가능성이 높습니다.  
        - PCA를 적용하면, 시험 점수 + 공부 시간을 조합한 새로운 주성분(PC1)이 생성됩니다.  
        - 이 주성분이 전체 데이터의 정보를 80% 설명한다면, 원래 변수를 유지할 필요 없이 PC1 하나로 핵심 정보를 분석할 수 있습니다.
    - 결국, PCA는 데이터 속 숨겨진 패턴을 찾고, 불필요한 정보를 줄여 더 간결한 분석을 가능하게 하는 도구입니다.

- center=T와 scale=T의 의미
    - center=T → 데이터의 평균을 0으로 조정
        - 각 변수의 평균을 빼서 모든 값이 평균 기준으로 재조정됨
        - 데이터의 위치를 이동시키지만, 변동성(분산)은 변하지 않음
        - 예: 키(170cm, 180cm)를 평균(175cm) 기준으로 재조정하면 (-5, +5) 형태로 표현됨
    - scale=T → 데이터를 표준화(standardization)함
        - 각 변수의 표준편차로 나누어 모든 값이 같은 범위를 가지도록 조정
        - 값들이 서로 다른 단위(예: 키 cm vs 몸무게 kg)일 때 중요
        - 표준화하면 데이터의 크기 차이가 줄어 PCA 결과가 더 안정적임
    - center=T와 scale=T을 적용하면 변수의 크기 차이를 조정하여 비교하기 쉽게 만든다는 의미예요!
- 분산이 크다는 의미
    - 분산(Variance)은 데이터가 얼마나 퍼져 있는지를 나타내요.
        - 쉽게 말해, 데이터 값들이 평균을 기준으로 얼마나 흩어져 있는지를 보여주는 개념입니다.
        - 예제: 학생 성적 분산
            - 학생 5명의 수학 점수가 (85, 90, 88, 92, 89)라면
                - 점수들이 평균(88.8) 근처에 몰려 있음 → 분산이 작음
            - 반면 다른 5명의 점수가 (60, 95, 70, 85, 100)라면
                - 점수들이 평균(82)에서 멀리 퍼져 있음 → 분산이 큼
        - 즉, 분산이 크면 데이터 값들이 평균 주변에 모이지 않고 넓게 퍼져 있다는 뜻입니다.
    - PCA에서는 첫 번째 주성분(PC1)의 분산이 가장 크다
        - PC1이 가장 많은 정보를 포함하고 있다는 뜻
            - PCA에서는 가장 많은 정보(즉, 데이터의 변동성)를 담고 있는 축을 첫 번째 주성분(PC1)으로 설정합니다.
            - 예를 들어, 학생들의 수학 성적, 과학 성적, 공부 시간 등의 데이터가 있을 때
                - PCA를 실행하면 가장 큰 변화를 설명하는 방향(축)이 찾아지고 → 이게 PC1이 됩니다.
            - 즉, PC1의 분산이 크다는 것은 PC1이 원래 데이터의 정보(변동성)를 가장 많이 포함하고 있다는 뜻이에요.
    - 데이터의 핵심 패턴을 가장 잘 설명하는 축이 PC1이므로 제일 중요한 주성분이 됩니다.
        - PCA는 데이터의 변화를 가장 잘 설명하는 방향을 찾아내는 알고리즘입니다.
            - 그래서 PC1은 데이터의 가장 중요한 패턴을 담고 있는 축이 됩니다.
            - 연구에서는 보통 PC1과 PC2를 사용하면 전체 데이터의 약 70~90%를 설명할 수 있다고 합니다.
        - 사례: 학생들의 국어, 수학, 과학 점수가 있다고 가정할게요.
            - 이 데이터를 PCA로 분석하면?
                - PC1 → 학생들의 학업 성취도를 대표하는 축 (시험 점수들의 공통 패턴)
                - PC2 → 학생들의 특정 과목 성향을 보여주는 축 (예: 수학을 유독 잘하거나 못하는 패턴)
            - 즉, PC1이 가장 많은 정보(변화)를 포함하고 있으므로 가장 중요한 주성분이 된다는 뜻이에요!
- 데이터의 80%를 설명한다는 의미
    - PCA를 적용하면 전체 데이터의 복잡한 정보를 몇 개의 주성분으로 표현할 수 있어요.
    - PC1 + PC2로 원래 데이터의 약 80%를 설명한다는 것은?
        - 원래 데이터의 변동성(정보량) 중 80%를 유지할 수 있다는 뜻
    - 즉, 모든 변수 대신 PC1과 PC2만 사용해도 데이터의 대부분을 분석할 수 있다는 의미예요!

- 주성분 분석 (이어서)
    - 차원(Dimension)은 데이터가 가지고 있는 변수(특성, features)의 개수를 의미해요.
        - 1차원 데이터 → 한 개의 변수만 있음 (예: 키만 기록된 데이터)
        - 2차원 데이터 → 두 개의 변수 (예: 키와 몸무게)
        - 3차원 데이터 → 세 개의 변수 (예: 키, 몸무게, 나이)
        - N차원 데이터 → 여러 개의 변수가 있는 고차원 데이터
        - 즉, 원래 데이터도 여러 개의 변수(특성)를 가지고 있다면 N차원 데이터라고 볼 수 있습니다.
    - PCA에서 차원 축소(Dimensionality Reduction)의 개념
        - PCA는 데이터의 중요한 정보를 유지하면서 차원을 줄이는 방법이에요.
        - 예를 들어, 원래 데이터가 5차원(5개의 변수)였다면 PCA를 통해 2~3차원으로 축소 가능
        - 하지만 데이터를 완전히 1차원으로 만들면 정보 손실이 많아 분석이 어려워질 수도 있음
        - PCA는 주성분의 개수에 따라 차원의 축소 정도가 달라집니다.
            - PC1 하나만 사용 → 1차원으로 축소됨
            - PC1 + PC2 사용 → 2차원으로 축소됨
            - PC1 + PC2 + PC3 사용 → 3차원으로 축소됨
            - 예제: RGB 이미지 압축
                - 원래 컬러 이미지는 3차원(RGB: 빨강, 초록, 파랑)
                - PC1만 사용하면 1차원이 되어 밝기 정보만 남을 가능성이 큽니다.
                - PC1 + PC2 사용하면 2차원이므로 여전히 색상 정보를 어느 정도 유지하면서도 축소됩니다.
                - PC1 + PC2 + PC3 사용하면 원래의 3차원과 동일하여 축소되지 않은 상태죠.

- 주성분 분석 이어서
    - 다차원 척도법 (MDS: MultiDimensionalScaling)
        - 데이터간 거리 정보의 근접성을 보존하는 방식으로 차원 축소해 시각화
        - 특징: 데이터 축소 목적, Stree값이 0에 가까울수록 좋음, xy축 해석이 불가
        - 종류
            - 계량적 mds: 양적척도 활용
            - 비계량적 mds: 순서척도 활용
    - 다차원 척도법은 데이터 간의 거리 정보를 최대한 유지하면서 차원을 줄여 시각화하는 기법이에요.
        - 쉽게 말하면, 고차원의 복잡한 데이터를 2차원이나 3차원으로 변환해 그래프 등으로 볼 수 있게 만드는 방법이죠.
        - 데이터 간 거리 정보 보존: 데이터를 단순히 축소하는 것이 아니라, 개별 데이터 간의 상대적인 거리(유사성 또는 차이)를 유지하면서 차원을 줄여요.
        - Stree 값: 변환된 차원에서 원래 거리 정보가 얼마나 잘 유지되었는지를 평가하는 값이에요. 0에 가까울수록 변환이 잘 된 것이죠.
        - xy축 해석 불가: 변환 후의 좌표 자체는 해석이 어렵고, 데이터의 상대적인 위치만 의미가 있어요.
        - 계량적 MDS: 데이터의 거리나 차이를 실수 값(연속적인 수치)으로 계산하는 방식.
