- K평균 군집화(k-means clustering)
    - 비계층적 군집화 방법, 거리기반
    - 특징
        - 안정된 군집 보장하나 최적의 보장은 어려움
        - 한번 군집에 속한 데이터는 중심점이 변경되면 군집 변할 수 있음
        - 초기 중심 값에 따라 결과 달라짐
    - 과정
        - 군집의 개수 K개 설정 (Elbow Methond로 최적의 K 설정)
        - 초기 중심점 설정
        - 데이터들을 가장 가까운 군집에 할당
        - 데이터의 평균으로 중심점 재설정
        - 중심점 위치가 변하지 않을 때까지 (가까운 군집 할당, 평균으로 중심점 재설정) 반복
    - K-medoids 군집화 = PAM
        - 평균 중심점 아닌 실제 데이터 중 하나인 대표(Medoids)를 설정
- 사례: 영화 추천 알고리즘
    - 스트리밍 플랫폼에서 사용자의 취향을 분석하고, 비슷한 취향을 가진 사람들을 그룹으로 묶어 맞춤형 추천을 제공한다고 가정해볼게요.  
    - K-평균 군집화 적용 과정  
        - 군집 개수(K) 설정  
            - 플랫폼은 사용자의 영화 취향을 분석해 5개의 군집(K=5) 으로 나누기로 결정했어요.  
            - (예: 로맨스 좋아하는 그룹, 액션 좋아하는 그룹, 공포 영화 좋아하는 그룹 등)  
        - 초기 중심점 설정  
            - 처음에는 임의의 사용자 몇 명을 각 군집의 대표(초기 중심점)로 설정합니다.  
            - 예를 들어, A 사용자는 로맨스 영화를 많이 봤으므로 "로맨스 군집"의 대표가 될 수 있어요.  
        - 사용자를 가장 가까운 군집에 할당  
            - 각 사용자의 영화 감상 패턴(좋아하는 장르, 시청 기록 등)을 분석한 후, 가장 가까운 중심점을 찾아 군집에 배정합니다.  
            - 예를 들어, B 사용자는 로맨스 영화 80%, 드라마 20%를 시청했다면 "로맨스 군집"으로 배정될 가능성이 크죠.  
        - 새로운 중심점 계산  
            - 각 군집의 평균을 계산해 새로운 중심점을 다시 설정합니다.  
            - 예를 들어, "액션 군집"은 처음엔 특정 영화들을 중심으로 형성되었지만, 데이터를 분석해보니 사용자의 취향이 '액션+SF'라면 새로운 중심점을 바꿀 수 있습니다.  
        - 중심점 변경 여부 확인, 반복  
            - 새로운 중심점을 기준으로 다시 군집을 할당하고 평균을 재설정하는 과정을 반복합니다.  
            - 중심점이 더 이상 변하지 않으면 최종 군집이 확정되죠.  
    - K-medoids 방식과 차이점  
        - K-medoids 방식에서는 중심점이 평균이 아니라 실제 사용자 중 한 명이 됩니다.  
        - 즉, "로맨스 군집"의 대표를 한 명의 실제 사용자로 설정하고, 그 사용자의 취향을 바탕으로 추천을 제공하는 것이죠.

`연관분석`
- 연관분석
    - 항목들간의 조건-결과로 이루어지는 패턴을 발견하는 기법 (장바구니 분석)
    - 특징
        - 결과가 단순하고 분명 (if~then~)
        - 강력한 비목적성 분석기법
        - 품목 수가 증가할수록 계산량이 기하급수적으로 증가
        - Apriori 알고리즘(최소지지도 활용 빈발항목집합 추출)을 활용 후, 연관분석울 수행
    - 순차패턴
        - 연관분석에 시간개념을 추가해 품목과 시간에 대한 규칙 찾는 기법
    - 연관분석(Association Analysis)이란?  
        - 연관분석은 항목들 간의 "함께 등장하는 패턴"을 발견하는 분석 기법이에요.
        - 특히, "이 항목을 구매한 고객은 저 항목도 살 가능성이 높다" 같은 패턴을 찾는 데 많이 사용되죠.  
        - 대표적인 예: 장바구니 분석  
            - 슈퍼마켓에서 사람들이 자주 함께 구매하는 상품을 분석하는 것이 대표적인 사례입니다.
            - 예를 들어, 많은 사람들이 "우유를 사면 빵도 함께 산다"는 패턴이 발견된다면, 우유를 진열할 때 빵을 가까운 곳에 배치하면 매출을 높일 수 있겠죠.  
    - 연관분석의 특징  
        - 조건 → 결과 형태의 패턴 분석 (if~then~ 구조)  
            - 예: “기저귀를 샀다면, 맥주도 살 가능성이 높다”  
        - 비목적성 분석기법 (목표가 정해져 있지 않음, 데이터를 통해 패턴을 찾아냄)  
        - 품목 수가 증가하면 계산량이 기하급수적으로 증가 (품목이 많아질수록 분석이 어려움)  
    - Apriori 알고리즘  
        - 연관분석을 수행할 때, 너무 많은 조합을 계산하면 비효율적이므로 Apriori 알고리즘을 사용합니다.  
        - 자주 함께 등장하는 항목들(빈발 항목 집합)을 먼저 추출한 후 연관분석을 수행하는 방식이죠.  
        - 이렇게 하면 불필요한 계산을 줄이고, 의미 있는 연관성을 빠르게 찾을 수 있어요.
    - 순차패턴 (Sequential Pattern)  
        - 연관분석에 시간 개념 을 추가한 기법입니다.  
        - 예를 들어, 고객이 먼저 스마트폰을 구매한 후 일정 시간이 지나면 보호필름을 구매할 확률이 높다면, 이를 활용해 마케팅 전략을 세울 수 있어요.  
        - 즉, “어떤 항목이 먼저 구매되고, 이후에 어떤 항목이 구매되는지”를 분석하는 방법이죠!

- 연관분석의 지표
    - 지지도
        - N(a교b)/전체=P(a교b)
        - a와 b 두 품목이 동시에 포함된 거래 비율
    - 신뢰도
        - P(a교b)/P(a)
        - a 품목이 거래될 때 b도 거래될 확률 (조건부확률)
    - 향상도
        - P(a교b)/{P(a)*P(b)}
        - a 품목과 b품목의 상관성
        - 향상도>1: 양의상관관계
        - =1: 상관없음
        = <1: 음의상관관계
- 연관분석은 독립성 가정을 따르지 않아요.
- 독립이라는 것은 한 사건의 발생이 다른 사건의 발생에 영향을 주지 않는다는 뜻인데,
- 연관분석은 특정 항목들이 함께 발생하는 패턴을 찾는 것이 목적이기 때문에 독립성을 가정하지 않죠.  
- 확률과 연관분석의 차이점  
    - 일반적인 확률에서는 두 사건 A와 B가 완전히 독립적이라면, 다음 식이 성립해야 합니다:  
        - P(A ∩ B) = P(A) × P(B)
        - 즉, 두 사건이 서로 영향을 주지 않는다면, 각각의 개별 확률을 곱한 값이 동시에 발생할 확률과 같아야 합니다.  
    - 하지만! 연관분석에서는 독립성을 확인하는 향상도(Lift)라는 개념을 사용해, 두 품목이 독립적인지 아닌지를 평가합니다.  
        - 향상도(Lift) = P(A ∩ B) / {P(A) × P(B)}  
            - 향상도 > 1 → A와 B는 양의 상관관계, 즉 함께 구매될 확률이 일반적인 확률보다 높음  
            - 향상도 = 1 → A와 B는 독립적, 즉 서로 영향을 주지 않음  
            - 향상도 < 1 → A와 B는 음의 상관관계, 즉 함께 구매될 확률이 일반적인 확률보다 낮음  
    - 예제: 장바구니 분석  
        - 사람들이 우유를 구매할 확률이 20%(P(A))이고, 빵을 구매할 확률이 15%(P(B))라고 해볼게요.  
        - 그런데 실제 데이터를 분석해보니, 우유와 빵을 함께 구매하는 확률이 10%(P(A ∩ B))라고 한다면?  
            - 향상도 =  P(A ∩ B) / {P(A) × P(B)} = 0.10 / (0.20 × 0.15) = 3.33  
            - 즉, 우유와 빵은 독립적이지 않고, 함께 구매될 확률이 일반적인 기대치보다 3.33배 더 높다는 것을 의미합니다!  
- 연관분석에서는 품목 간 관계를 찾기 위해 확률의 독립성 가정을 깨고, 향상도(Lift)를 통해 상관관계를 측정합니다.  
- 그래서 단순한 확률 곱셈이 아니라, 실제 데이터를 활용해 품목 간의 관계를 분석하는 방식이죠.

- 맥주를 구매할 때 치킨 구매 확률에 대한 신뢰도와 향상도

|거래코드|품목|거래횟수|
|:-:|:-:|:-:|
|1|맥주|10|
|2|치킨|20|
|3|햄버거|70|
|4|맥주,치킨|20|
|5|맥주,햄버거|30|
|6|치킨,햄버거|10|
|7|맥주,치킨,햄버거|40|

- 10+20+70 +20+30+10+40 = 200
- 맥주 구매 확률 = 10+20+30+40을 200으로 나눈다 = 100/200 = 50/100 = 0.5
- 치킨 구매 확률 = 20+20+10+40을 200으로 나눈다 = 90/200 = 45/100 = 0.45
- 맥주,치킨의 지지도 = 20+40을 200으로 나눈다 = 30/100 = 0.3
- 맥주->치킨의 신뢰도 = 0.3/0.5 = 0.6
- 맥주,치킨의 향상도 = 0.3/(0.5*0.45) = 300/(5\*45) = 3 25 4 / 9 25 = 4/3 = 1.33..
    - 맥주 치킨 향상도 1보다 크니까 양의 상관관계 가진다.
