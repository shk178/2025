1. R기초와 데이터 마트
`데이터마트`
- 데이터마트dm
    - 데이터웨어하우스 한 분야, 특정목적 위해 사용 = 소규모 데이터웨어하우스
- 요약변수, 파생변수

`결측값과 이상값 탐색`

    - 데이터의 의미 찾기 위해 통계/시각화로 파악

        - 저항성의 강조 = 자료 변동에 민감하지 않음

        - 그래프를 통한 현시성 = 시각화 통한 효율적 파악
- 결측값 처리: 존재하지 않는 데이터를 null/na로 표시

    - 다중대치법: 여러 번 대치 (대치->분석->결합)

2. 통계분석
`통계학 개론`
- 전수조사와 표본조사
    - 전수조사: 전체를 다 조사, 시간과 비용 많이 소모
    - 표본조사: 일부만 추출해 모집단 분석
        - 모집단->표본추출로 표본집단
        - 표본집단에서 가설검정(확률분포 타당성)으로 모집단 예측
- 자료의 척도 구분
    - 질적척도
        - 명목척도: 어느 집단에 속하는지 나타내는 자료 (대학교, 성별)
        - 순서척도(서열척도): 서열관계 존재하는 자료 (학년, 순위)
    - 양적척도
        - 등간척도(구간척도): 구간 사이 간격이 의미가 있으며 덧셈 뺄셈만 가능 (온도, 지수 등)
        - 비율척도: 절대적 기준 0이 존재하고 사칙연산 가능한 자료 (무게, 나이 등)
- 확률적 표본 추출 방법
    - 랜덤 추출법: 무작위로 표본 추출
    - 계통 추출법: 데이터에 번호 부여해 일정 간격으로(1번 뽑으면 3번, 5번...) 추출
    - 집락 추출법(군집 추출법)
        - a아파트, b아파트, c아파트 다 조사하려면 시간 걸린다. a아파트만 가서 추출한다.
            - a아파트가 b c 아파트 대변할 수 있다. abc가 동질적인 특징 있어서..
            - 아파트 내에서는 다양한 데이터=이질적인 데이터 있다.
        - 여러 군집으로 나눈 뒤 군집 선택해 랜덤 추출
        - 군집 "내" 이질적 특징/군집 "간" 동질적 특징
    - 층화 추출법
        - 1학년 2학년 3학년이 있다. 1학년 100명 2학년 200명 3학년 300명 있다.
            - 1학년에서만 12명 뽑으면 학교 전체 대변 불가..
            - 층을 나눠서 1학년 2학년 3학년 몇 명씩 뽑는다.
            - 1학년 100명 간에 동질적.. 1학년과 2학년은 이질적
        - 군집 "내" 동질적 특징/군집 "간" 이질적 특징
        - 같은 비율로 추출 시 -> 비례 층화 추출법
            - 1학년 2학년 3학년 1:2:3으로 뽑는다면 비례 층화다.
    - 복원/비복원 추출법
        - 복원 추출: 추출되었던 데이터를 다시 포함시켜 표본 추출
        - 비복원 추출: 추출되었던 데이터는 제외하고 표본 추출
- 비확률적 표본 추출 방법
    - 편의 추출법: 연구자가 쉽게 접근 가능한 대상으로 표본 추출
    - 의도적 추출법: 연구자가 특정 기준 정하고 이에 맞는 표본 추출
    - 할당 추출법: 특정 기준으로 나눈 후 그 그룹에서 할당된 수만큼 추출
    - 눈덩이 추출법: 초기 응답자로부터 새로운 응답자 추천 받는 방식
    - 자기선택 추출법: 응답자가 스스로 조사에 참여할지 결정

- 기초 통계량
    - 중심경향성 측면
        - 산술평균
            - 일반적인 평균 개념으로, 모든 값을 더한 후 데이터 수로 나눈값
        - 기하평균
            - 모든 값들 곱하고 n제곱근 구한다..비율적 증가율
        - 조화평균
            - 역수의 산술평균 구한 후 다시 역수 취한다..비율 계산
        - 중앙값
            - 데이터를 크기 순서로 나열해 중간에 위치한 값
        - 최빈값
            - 데이터에서 가장 자주 나타나는 값
    - 분산 정도 측면
        - 분산
            - 각 데이터가 평균과 얼마나 떨어져 있는지 나타내는 지표
        - 표준편차
            - 분산에 제곱근 취한 값
        - 사분위수(IQR)
            - 데이터 상위 75%와 하위 25%의 중간 범위
    - 관계 측면 (상관관계=전후양상 보고판단)
        - 공분산: 두 확률 변수의 상관정도
            - 공분산이 0: 전혀 상관x
            - 공분산 양수: 양의 상관관계
            - 공분산 음수: 음의 상관관계
            - 공분산은 무한대 -무한대까지 나온다. 얼마나 강한지약한지 판단 어렵
        - 상관계수: 상관정도를 -1~1값으로 표현
            - 상관계수 1: 정비례관계
            - 상관계수 0: 상관x
            - 상관계수 -1: 반비례관계
            - 1부터 -1이라서 강약 판단이 쉽다.
        - 공분산과 독립성의 관계
            - 두 변수가 독립이면 공분산은 0
            - 공분산이 0이라고 두 변수가 독립은x (x-y 2차함수인데 공분산 0)
- 첨도와 왜도
    - 첨도: 자료분표가 얼마나 뾰족한지 척도
        - 첨도=3일 때, 정규분포 형태 (3을 빼서 0 기준으로 정규분포 판단하기도)
        - 첨도 클수록 뾰족
        - 넓이는 같다.
    - 왜도: 자료분포 비대칭 정도 (0이면 대칭)
        - 왜도 0일 때: 중앙값, 평균값, 최빈값이 같은 것은 맞습니다. 데이터가 완벽하게 대칭 분포를 이루면 이렇게 됩니다.
        - 중앙값: 넓이를 두 부분으로 나누고, 일반적으로 경사가 완만한 쪽으로 이동하는 경향이 있습니다. 하지만, "가파른 쪽에는 경계에 최빈값 말고 아무 값도 없다"는 표현은 조금 조정이 필요할 수도 있습니다. 실제로 데이터 분포에 따라 달라질 수 있기 때문입니다.
        - 평균값: 일반적으로 꼬리 방향으로 끌려가는 것이 맞습니다. 특히, 왜도가 있을 경우 평균값은 중앙값보다 더 최빈값에서 멀어지게 됩니다. 즉, 최빈값 → 중앙값 → 평균값 순으로 정렬될 가능성이 높습니다.
        - 최빈값: 분포에서 가장 빈도수가 높은 지점을 의미하며, 일반적으로 분포의 가장 뾰족한 부분에 위치합니다.
        - 왜도>0: 왼쪽이 가파르고 오른쪽이 완만함
            - "최빈값"<중앙값<평균값
        - 왜도<0: 왼쪽이 완만하고 오른쪽이 가파름
            - 평균값<중앙값<"최빈값"

- R Summary(통계요약) 함수 결과 해석
    - Age 변수: 이 변수는 나이를 나타내며, 요약된 통계 값은 다음과 같습니다.
        - Min (최솟값): 0.17 (가장 어린 사람의 나이)
        - 1st Qu (제 1사분위수): 21.00 (하위 25%에 해당하는 나이)
        - Median (중앙값): 27.00 (전체 나이 분포에서 중앙에 위치한 값)
        - Mean (평균값): 30.27 (전체 나이의 평균)
        - 3rd Qu (제 3사분위수): 39.00 (상위 25%에 해당하는 나이)
        - Max (최댓값): 76.00 (가장 나이가 많은 사람)
        - NA's (결측값 개수): 86 (결측치가 86개 존재함)
        - 즉, Age 변수는 0.17세부터 76세까지의 범위를 가지며, 중앙값이 27세이고 평균값은 30.27세입니다.
    - Embarked 변수 (특정 승선 항구 여부)
        - 0: 320 → 320명은 특정 항구에서 승선하지 않음
        - 1: 98 → 98명은 특정 항구에서 승선함 (예: Southampton에서 승선한 경우)
        - Embarked 변수는 원래 C(Q), S(Q), Q(Q) 등 여러 항구 코드를 포함하는 범주형 변수인데, 이를 특정 기준("Southampton에서 탑승했는가?")으로 변환해 이진 변수처럼 표현할 수도 있습니다.
    - Age와 Embarked 변수가 같은 데이터 집단에 속할 수 있습니다. 예를 들어, 여객선 승객 데이터처럼 여러 개의 변수(Age, Embarked, Gender 등)가 같은 사람들의 정보를 나타낼 수 있죠. 즉, 각 변수가 개별적인 특성을 나타내지만, 같은 승객들에 대한 정보를 저장하고 있을 가능성이 높습니다.
        - Age 변수에서 결측치가 아닌 데이터 개수: Age 변수에서 결측치(NA)가 86개라고 주어졌다면, 전체 데이터에서 결측치가 아닌 값(즉, 실제 데이터가 존재하는 값)의 개수를 계산할 수 있습니다.
            - 전체 데이터가 418개라고 가정하면: {총 데이터 개수} - {결측치 개수} = 418 - 86 = 332
            - 즉, Age 변수가 존재하는 데이터 개수는 332개입니다.
            - 요약 통계에는 NA's 86이 표시되지만, 결측치가 아닌 값은 따로 표시되지 않는 경우가 많습니다.
        - 네, 맞아요! Embarked 변수의 값을 보면 승객 수가 총 418명(320명 + 98명)이라는 정보를 얻을 수 있습니다. 이를 통해 데이터 전체 크기를 추정할 수 있었던 거죠.
    - age 변수는 mean과 median 등 존재한다: 수치형 변수이다.
        - median중앙 < mean평균: 중앙평균이면 오른쪽이 더 완만, 왜도가 양수
            - 왜 왜도가 양수이면 오른쪽이 더 완만할까?
                - 평균값이 꼬리를 따라 이동 → 데이터 분포에서 평균은 극단값(꼬리)의 영향을 많이 받습니다. "왜도가 양수일 때 평균은 오른쪽으로 이동하기 때문에" 오른쪽으로 길게 퍼지면서 완만해집니다.
                - 최빈값(가장 많이 등장한 값)은 왼쪽에 위치 → 왜도가 양수인 경우, 데이터의 대부분은 상대적으로 낮은 값(왼쪽)에 몰려 있고, 오른쪽에 몇 개의 높은 값(극단값)이 존재합니다. 이는 오른쪽 꼬리가 길고 완만해지는 원인이 됩니다.
                - 경사가 급한 곳은 왼쪽, 완만한 곳은 오른쪽 → 왼쪽에는 데이터가 많이 밀집해 있고 변화가 빠르게 일어나므로 경사가 가파릅니다. 반대로 오른쪽은 극단값들이 천천히 감소하는 형태라 경사가 완만해집니다.
    - embarked 변수는 집단의 빈도수를 나타낸다: 범주형 변수다.
        - 범주형 변수는 크게 이진(Binary) 변수와 다항(Multinomial) 변수로 나눌 수 있어요.
        - 순서가 있는 범주형 변수는 서열(Ordinal) 변수라고도 해요. 예를 들면 고객 만족도(매우 만족, 만족, 보통, 불만족)처럼 순서가 중요한 경우죠.
